مقدمه
</div>
    بهینه‌سازی یک راه مهم زندگی کردن است و ما همیشه در کار‌هایی که می‌کنیم منابع محدودی داریم و در استفاده از آنها مجبور به انتخاب بهینه هستیم.
    بهینه‌سازی به تنهایی یک موضوع بسیار گسترده است و یکی از مفاهیم مهم آن برنامه‌ریزی خطی (Linear Programming) است.  
    <hr>
    <div style="font-size: 18px; color: #ff6600; font-weight: bold">
        برنامه‌ریزی خطی
    </div>
    برنامه‌ریزی خطی چیست؟ برنامه ریزی خطی یک تکنیک ساده است که با استفاده از آن روابط پیچیده را از طریق تابع‌های خطی نمایش می‌دهیم و در نهایت نقطه بهینه این دستگاه را پیدا می‌کنیم.
    نکته مهم در متن قبلی این است که کلیت روابط ممکن است بسیار پیچیده باشد ولی ما با برنامه‌ریزی خطی آنها را بسیار ساده و جامع مثال می کنیم.
   <br>
    استفاده و کاربرد برنامه‌ریزی خطی در همه‌جا دیده می‌شود برای مثال شما وقتی می‌خواهید تمارین خود را به گونه‌ای و به ترتیبی انجام دهید که به موقع تمامی آنها را پیش از ددلاین انجام داده باشید از برنامه‌ریزی خطی به صورت ناخودآگاه بهره می‌برید. حتی هنگامی که شما برای بازگشت از دانشگاه به خانه مسیریابی می‌کنید در عمل از این تکنیک بهره می‌برید.
    <br>
    در بخش بعد به بیان چند مسئله مهم در این زمینه می‌پردازیم.
</div>

طرح چند مسئله

    <div style="direction:rtl;font-size: 18px; color: #ff6600; font-weight: bold">
        مسئله اول: پستچی
    </div>
     <div>
         پستچی داریم که می‌خواهد ۶ بسته را در روز به مقصد برساند پست‌خانه در نقطه A قرار دارد و ۶ مقصد با نقاط (U,V,W,X,Y,Z) در شکل نمایش داده شده است اعداد روی خطوط فاصله بین خانه‌ها را نشان می‌دهد برای ذخیره زمان و بنزین پستچی می‌خواهد کوتاهترین مسیر ممکن را طی کند.
        <img src="img/pic1.png" />
     </div>
    <div>
     برای این مسئله پستچی تمامی مسیر ها را چک می‌کند تا بتواند بهترین آنها را انتخاب کند این تکنیک انتخاب همان برنامه‌ریزی خطی است.در این مثال هدف پستچی رساندن همه نامه‌ها به موقع است. برنامه‌ریزی خطی برای پیدا کردن بهترین جواب با وجود شروط خطی روی ورودی‌ها استفاده می‌شود.
    <br>
    در مثال بعدی به بیان دقیق‌تر شروط و معادلات می‌پردازیم:
    <hr>
    <div style="direction:rtl;font-size: 18px; color: #ff6600; font-weight: bold">
        مسئله دوم: برنامه‌ای برای زندگی احسن!!
    </div>
    <div>
               می‌دانیم هر هفته از ۱۶۸ ساعت تشکیل شده‌است فرض کنید شما برای یک زندگی سالم و مفید می‌خواهید زمان خود را به سه بخش درس و دانشگاه (S) ، تفریح (P) و یا دیگر کارها مانند خوابیدن و غذا خوردن (E) است.
        برای اینکه سلامت روان خود را نگه داریم نیاز داریم 
        <br>
        (P + E &ge; 70) 
        همچنین برای گرفتن نمرات خوب از دروس خود نیز باید (S &ge; 60) 
         و همچنین شرط اول برای شاداب بودن در طول روز ۸ ساعت خوابیدن است. و همچنین ما اگر زیادی تفریح کنیم یا باید بیشتر بخوابیم یا بیشتر درس بخوانیم.
        2S + E -3P &ge; 150
        </div>
        <div style="direction:rtl;font-size: 18px; color: #ff6600; font-weight: bold">
            پرسش اول:
        </div>
        <div>
            آیا ما می‌توانیم این کار را انجام دهیم آیا جوابی وجود دارد؟
        </div>
        <div>
            بله برای مثال:
            S=80,P=20,E=68
        </div>
        <div style="direction:rtl;font-size: 18px; color: #ff6600; font-weight: bold">
            پرسش دوم:
        </div>
        <div>
            فرض کنید مقداری خوشحال بودن ما با عبارت (‌2P + E) بیان شود جوابی را پیدا کنید که در روابط صدق می‌کند. 
            <br>
            به فرمول بالا که می‌خواهیم آنرا بیشینه(گاهی اوقات کمینه) کنیم تابع هدف می‌گویند
        </div>
        رابطه بالا نیز برنامه‌ریزی خطی است زیرا تمام شروط ما خطی و همچنین تابع هدف نیز خطی است.
           
</div>

واژه‌شناسی در برنامه‌ریزی خطی:
</div>
<ul>
    <li>متغیرهای تصمیم‌گیری(Decision Variables): متغیرهایی هستند که خروجی ما را مشخص می‌کنند. برای حل هر کدام از مسایل نیاز داریم این متغیر ها را مشخص کنیم </li>
    <li>تابه هدف(Objective function): به عنوان هدف تصمیم‌گیری‌های ما است</li>
    <li>شروط و  (Constraints): محدودیت‌ها روی متغیرهای تصمیم‌گیری است.</li>
    <li>محدودیت‌های نامنفی(Non-negative restriction):برای همه برنامه‌ریزی‌های خطی  همه‌ی متغیر‌ها باید نامنفی باشند.</li>
</ul>
    در بخش بعدی به بیان فرم استاندارد می‌پردازیم.
</div>

<div id="section4" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 22px; color: #ff6600; font-weight: bold">
فرم استاندارد:
</div>
<div>
    در تعریف برنامه‌ریزی خطی پیش از این بیان شد که (LP) روشی برای رسیدن به خروجی بهینه با تصدیق در دستگاهی از شروط که دارای رابطه خطی هستند، می‌باشد.اما اگر بخواهیم دقیق‌تر بیان کنیم برنامه‌ریزی خطی می‌تواند مسئله موجود که بیشینه یا کمینه کردن یک تابع هدف هست را در وجود تعدادی محدودیت خطی حل کند.
<br>
اما فرم استاندار چیست؟ در حالت کلی فرم استاندارد متشکل است از:
</div>
    <div style="direction:rtl;">
    متغیرها: $$ x=(x_1, x_2,..,x_d) ^T $$
    <br>
    تابع هدف (n): $$c.x$$ 
    <br>
    شروط: $$ Ax \le b \rightarrow A = (n \times d) $$ 
    </div>
    و ما در نهایت تابع هدف را بیشینه می‌کنیم. 
    <br>
     LP کاربرد های بسیاری دارد که از نمونه آن می‌توان به شار و پیدا کردن کوتاهترین مسیر بین یک راس مشخص با دیگر رئوس در گراف (shortetst path) اشاره کرد (یعنی به سادگی می‌توان این سوال‌ها را به  LP تبدیل کرد و سپس آنها را حل کرد). 
</div>
<br><hr>

<div id="section5" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 22px; color: #ff6600; font-weight: bold">
قضیه دوگان:
</div>
    این قضیه بیان می‌کند برای هر مسئله که به شکل استاندارد (LP) بیان می‌شود:
    <div style="direction:ltr;">
        Maximize: $ c.x $
        <br>
        Subject to: $ Ax \le b, x \ge 0 $
    </div>
    یک دوگان وجود دارد که:
    <div style="direction:ltr;">
        Minimize: $ b.y $
        <br>
        Subject to: $ A^T y \ge c , y \ge 0 $
    </div>
    این خاصیت اهمیت بسیاری دارد و می‌تواند قضایا زیادی را اثبات کند مانند اینکه اگر قضیه ( Max flow) را به صورت برنامه‌ریزی خطی بیان کنیم از روی این خاصیت ثابت می‌شود جواب آن، جواب دوگانش است که دوگانش هم قضیه (Min cut) خواهد شد.
</div><hr>

<div id="section6" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 22px; color: #ff6600; font-weight: bold">
تبدیل به فرم استاندارد:
</div>
<div>
    ممکن است پس از نوشتن مسئله خود به صورت برنامه‌ریزی خطی شکل آن به شکل استاندارد نباشد ولی راه های بسیار ساده‌ای برای تبدیل هر دستگاه معادله‌ای به فرم استاندارد وجود دارد. در زیر به بیان این موارد می‌پردازیم:
</div>
    <ul>
        <li>کمینه کردن تابع هدف: برای اینکار ضرایب را در -۱ ضرب کنید و تابع هدف بدست آمده را بیشینه کنید</li>
        <li>متغیری وجود دارد که شرط نامنفی بودن ندارد:
            <div style="direction:ltr">
        $ x_i \rightarrow x_j - x_k$ و $x_j \ge 0, x_k \ge 0 $
            </div>
        </li>
        <li>شروطی که تساوی دارند:
             <div style="direction:ltr">
        $ x = b \rightarrow x\le b , x\ge b $
            </div>
             یعنی در حقیقت به جای معادله تساوی، دو تامساوی بالا را به دستگاه معادلات اضافه می‌کنیم
.        </li>
        <li>شروطی که به شکل بزرگتر یا مساوی هستند:
        <div style="direction:rtl">
            .ضرایب را در -۱ ضرب می‌کنیم و تبدیل به کوچکتر مساوی می‌کنیم
         </div>
        </li>
    </ul>
    نکته مهم این است که تعاریف بالا برای فرم استاندارد گفته شده و نه حالت دوگان البته حالت دوگان نیز به سادگی قابل تفهیم است.
</div><hr>

الگوریتم سیمپلکس
</div>
<div>
پس از مطرح کردن کلیت برنامه‌ریزی به الگوریتم‌های حل آن می‌رسیم، می‌توان اثبات کرد که یک مسالهٔ برنامه‌ریزی خطی را می‌توان در زمان چندجمله‌ای حل کرد؛ اما در این قسمت ما به معرفی الگوریتم سیمپلکس می‌پردازیم که در worst-case مساله را در زمان اجرای نمایی حل می‌کند اما در عمل خیلی بهتر از زمان نمایی عمل می‌کند.

مقدمه‌ای بر کلیت کار سیمپلس
