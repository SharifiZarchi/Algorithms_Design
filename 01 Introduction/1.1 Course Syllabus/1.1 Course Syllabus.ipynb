سیاست تاخیر
</div>     
    شما در این درس مجاز هستید که تمرین‌های عملی را با تاخیر بفرستید. به ازای هر ساعت تاخیر، سه درصد از نمره‌ی آن سوال کسر می‌شود. برای تمرین‌های تئوری امکان ارسال با تاخیر وجود ندارد.
<br>
    هم‌چنین شما در این درس ۱۲۰ ساعت کوپن تاخیر برای مجموعه‌ی تمرین‌های تئوری و عملی در اختیار دارید و میتوانید بدون کسر نمره از آن استفاده کنید و تمرین خود را دیرتر از ددلاین بفرستید. دقت کنید که این کوپن‌ها روی کل تمرین اعمال می‌شوند و برای مثال در یک تمرین تئوری شما می‌توانید ددلاین کل تمرین را ۲ ساعت عقب بیاندازید؛ درحالیکه ارسال با تاخیر که با کسر نمره همراه است برای هر سوال به صورت تکی اعمال می‌شود.
<br><br>
    <strong>
    نکته:
    </strong>
    با وجود اینکه شما مجموعا ۱۲۰ ساعت (معادل ۵ روز)
    کوپن تاخیر دارید،
     برای یک تمرین (عملی یا تئوری) نهایتا می‌توانید از ۷۲ ساعت (معادل ۳ روز) کوپن استفاده کنید.
    <br>
    کوپن‌های تاخیر در پایان ترم در یک فرم جمع‌آوری خواهند شد و نمرات مربوطه اصلاح می‌شوند.

سیاست تقلب
</div>     
در صورت کشف تقلب از شما در هر تمرین تئوری یا عملی، در مرتبه‌ی نخست نمره‌ی آن تمرین برای شما منفی ۲۰۰ در صد در نظر گرفته می‌شود و در مرتبه‌ی دوم تقلب شما به کمیته‌ی آموزش ارجاع داده خواهد شد و ضمن ثبت در سوابق تحصیلی، آن کمیته در مورد چگونگی جریمه تصمیم خواهد گرفت. اگر ارسال شما مشکوک به تقلب باشد، این مورد از طریق ایمیلی که در سایت edu وارد کرده اید به اطلاع شما خواهد رسید؛ بنابراین از درست بودن و چک‌کردن مرتب این ایمیل اطمینان حاصل کنید.
<br><br>
