<div style="font-size: 22px; color: #ff6600; font-weight: bold">
مقدمه
</div>
<ul>
ضرب‌المثلی قدیمی در میان مهندسان وجود دارد که می‌گوید: "ارزان، سریع، مطمئن، دو تا را انتخاب کن!"
<br>
در دنیای مسائل علوم کامپیوتر هم وضعیت مشابهی برقرار است.  همانطور که در فصل‌های پیش آموختیم، بسیاری از مسئله‌های بهینه‌سازی معروف
NP Hard
هستند. بنابراین اگر 
$P \neq NP$
نمی‌توانیم الگوریتم‌هایی داشته باشیم که در زمان چند‌جمله‌ای جواب بهینه یک مسئله را برای هر نمونه از آن به ما بدهند. بنابراین وقتی با این مسائل در کاربرد‌های واقعی سر و کار داریم، حداقل یکی از این شرط‌ها را باید کنار بگذاریم. 
<br>
یکی از راه‌ها این است که شرط چند‌جمله‌ای بودن الگوریتم را کنار بگذاریم و مانند الگوریتم‌های 
$A*$
یا 
CSP
فضای جواب‌های مسئله را جستجو کنیم و بر اساس یک آستانه‌ی مناسب جواب نیمه اپتمیال را به دست آوریم. اما ممکن است مجبور شویم برای به دست آوردن جواب قابل قبول زمان نامشخصی را بسته به شرایط مسئله صرف کنیم که در خیلی از کاربرد‌ها مناسب نیست. 
<br>
راه دیگر این است که شرط "برای هر نمونه" را در نظر نگیریم و با در نظر گرفتن شروطی دیگر، حالت خاصی از مسئله را حل کنیم. اما این راه نیز در صورتی جواب می‌دهد که کاربردمان در این حالت‌های خاص بگنجد که خیلی از اوقات رخ نمی‌دهد. 
<br>
راه سوم که مرسوم‌ترین راه است، این است که به جای یافتن جواب اپتیمال، جوابی را پیدا کنیم که "به اندازه‌ی کافی خوب" است. الگوریتم‌های تقریبی نیز از همین جا به وجود آمده‌اند. در این الگوریتم‌ها ما به دنبال جوابی می‌گردیم که با نسبت خوبی به ازای هر نمونه از مسئله، جواب اصلی را تقریب بزند. 
</ul>    
<div id="section1.1">
<div style="font-size: 17px; color: #ff6600; font-weight: bold">
کاربرد الگوریتم‌های تقریبی
</div>
<ul>     
با توجه به توضیحات بخش قبل در می‌یابیم که الگوریتم‌های تقریبی وقتی به کار می‌روند که یافتن جواب بهینه‌ی مسئله غیر‌عملی است. البته باید توجه کنیم که بسیاری از مسائل دنیای واقعی شروط بسیار پیچیده‌ای دارند که طراحی یک الگوریتم تقریبی را برایشان سخت می‌کند. در اینجا ما مجبور می‌شویم مسئله را ساده‌تر کنیم و بسیاری از شروط را سست کنیم  تا بتوانیم به الگوریتمی مناسب برسیم. سپس از روی این الگوریتم برای حالت ایده‌آل، الگوریتمی برای حالت کلی با تقریب مناسب پیدا می‌کنیم. (مثالی از این ساده سازی شروط را در بخش بعد خواهیم دید)  
</ul>         
<hr>

<div id="section1.2" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 17px; color: #ff6600; font-weight: bold">
تعاریف و ضریب تقریب
</div>
<ul>
اکنون که متوجه شدیم چرا و در کجا از الگوریتم‌های تقریبی استفاده می‌کنیم، در این بخش به بررسی تعاریفی که در حل مسائل با این الگوریتم‌ها به کار می‌آید می‌پردازیم.
<br>
الگوریتم‌های تقریبی برای حل مسائل بهینه‌سازی به کار می‌رود که  
NP Hard
هستند. این مسائل می‌تواند مسئله‌ی مینیمم‌سازی یا ماکسیسم‌سازی باشد.  
<br>
<span style="color: red">
تعریف:
</span>
یک الگوریتم 
$\alpha-approx$
برای یک مسئله‌ی بهینه سازی الگوریتمی با زمان چند‌جمله‌ای است که به ازای هر نمونه از این مسئله جوابی تولید می‌کند که مقدار آن در نسبت 
$\alpha$
برابر جواب بهینه مسئله است. 
<br>
در تعریف فوق مقدار $\alpha$
را
<span style="color: red">
ضریب تقریب
</span>
یا ضمانت عملکرد الگوریتم
می‌نامیم. از تعریف مشخص است که برای مسائل مینیمم‌سازی داریم 
$\alpha > 1$
و برای مسائل ماکسیمم‌سازی  
$\alpha < 1$
(چرا؟)
بنابراین یک الگوریتم 
$\frac{1}{2}-approx$
برای یک مسئله ماکسیمم‌سازی الگوریتمی از زمان چندجمله‌ای است که همواره جوابش حداقل نصف جواب بهینه مسئله است.
مشابها یک الگورتیم 
$3-approx$
برای یک مسئله مینیمم‌سازی الگوریتمی از زمان چند‌جمله‌ای است که جوابش به ازای هر نمونه مسئله حداکثر 3 برابر جواب بهینه است.
<br>
در ادامه‌ی این دفترچه به مثال‌هایی جالب از طراحی این الگوریتم‌ها برای مسائل 
NP Hard
معروف می‌پردازیم.
</ul>
</div> 

<div id="section2" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 22px; color: #ff6600; font-weight: bold">
پوشش رأسی
</div>
<div id="section2.1">
<div style="font-size: 17px; color: #ff6600; font-weight: bold">
معرفی مسئله
</div>
<ul>
مسئله پوشش راسی (Vertex Cover) که ما در اینجا حالت وزن‌دار آن را بررسی می‌کنیم، یک مسئله کلاسیک 
NP Complete 
است که الگوریتم‌های تقریبی گوناگونی برای آن ارائه‌شده‌است. اکنون به تعریف مسئله می‌پردازیم. 
<br> 
گراف بدون جهت 
$G = (V, E)$
داده‌شده‌است به طوری که به هر راس
$i \in V$
وزن
$w_{i} \ge 0$
نسبت داده‌شده‌است.
می‌خواهیم مجموعه‌ی
$C \subseteq V$
از رئوس را انتخاب کنیم به گونه‌ای که مجموع وزن رئوس 
$C$
کمینه شود و همچنین برای هر یال 
$(i, j) \in E$
داشته باشیم
$i \in C$
یا 
$j \in C$.
یعنی یکی از دو سر هر یال حتما در 
$C$
بیاید.
<br> 
برای مثال شکل زیر یک پوشش راسی مینیمال را نشان می‌دهد.
<img src="img/vc.png" width = 400 />
<br> 
</ul>
</div>

<div id="section2.2" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 17px; color: #ff6600; font-weight: bold">
راه‌حل با برنامه‌ریزی خطی
</div>
<ul>
در فصل‌های پیش یاد گرفتیم که برنامه‌ریزی خطی یک تکنیک کاربردی برای نمایش و حل مسائل بهینه‌سازی است. بنابراین ابتدا سعی می‌کنیم مسئله‌ی فوق را با برنامه‌نویسی خطی فرموله کنیم.
<br> 
برای هر راس 
$i \in V$
اگر 
$i$
عضو مجموعه‌ی انتخاب شده باشد، 
متغیر 
$x_{i}$
را برابر 1 و در غیر این صورت برابر 0 در نظر می‌گیریم.  
بنابراین به سادگی می‌بینیم که مسئله‌ی پوشش راسی را می‌توان به صورت زیر بیان کرد:
$$min \sum_{i} x_{i}w_{i}$$
$$s.t. \quad\forall (i, j) \in E  : x_{i} + x_{j} \ge 1 $$
$$\forall i \in V: x_{i} \in \{0, 1\} $$    
<br> 
بنابراین این مسئله را شکل یک برنامه‌ریزی خطی با اعداد صحیح در آوردیم. اما می‌دانیم که برنامه‌های خطی صحیح 
(integer linear programs)
<br>
NP Hard
هستند. 
<br> 
بنابراین برای این که بتوانیم مسئله‌ی فوق را با راه حل چند‌‌جمله‌ای به صورت تقریبی حل کنیم شرط صحیح بودن 
$x_{i}$
ها را بر می‌داریم و به جای آن شرط
$x_{i} \in [0, 1]$
می‌گذاریم. با یک ساده‌سازی دیگر برنامه خطی زیر به دست می‌آید
$$min \sum_{i} x_{i}w_{i}$$
$$s.t. \quad\forall (i, j) \in E  : x_{i} + x_{j} \ge 1 $$
$$\forall i \in V: x_{i} \ge 0 $$ 
حال به برنامه‌ای خطی رسیدیم که در زمان چند‌جمله‌ای قابل حل است! اکنون نیاز به روشی داریم که پاسخ این مسئله را به مسئله‌ی اصلی‌مان تبدیل کند. فرض کنیم 
$x^{*}$
جواب 
LP
فوق است. مجموعه 
$C$
را به صورت زیر تعریف می‌کنیم
$$C = \{i | x^{*}_{i} \ge \frac{1}{2}\}$$
ادعا می‌کنیم
$C$
یک پوشش راسی است.
برای هر یال 
$(i, j) \in E$
طبق شرط 
LP
داریم
$x^{*}_{i} + x^{*}_{j} \ge 1$
بنابراین یا 
$x^{*}_{i} \ge \frac{1}{2}$
و یا 
$x^{*}_{j} \ge \frac{1}{2}$
.
اما این نشان می‌دهد که حداقل یکی از 
$i$
و 
$j$
در
$C$
هستند. پس  طبق تعریف
$C$
یک پوشش راسی است.
<br> 
همچنین اگر 
$C^{*}$
پوشش راسی مینیمال باشد داریم
$$w(C^{*}) = \sum_{i \in C^{*}} w_{i} = OPT_{ILP} \ge OPT_{LP} = \sum_{i} x^{*}_{i}w_{i} \ge \frac{1}{2} \sum_{i \in C} w_{i} = \frac{1}{2}w(C)$$
$$ \Rightarrow w(C) \le 2w(C^{*})$$
که در اینجا مجموع وزن رئوس 
$C$
را با 
$w(C)$
نشان دادیم و 
$OPT_{ILP}$
و
$OPT_{LP}$
به ترتیب جواب اپتیمال برنامه خطی اول و دوم هستند.
<br> 
<span style="color: red">
سوال:
</span>
چرا داریم
$OPT_{ILP} \ge OPT_{LP}$
؟
<br> 
بنابراین جواب بهینه مسئله که از روی 
LP
ساختیم  حداکثر دو برابر جواب بهینه مسئله اصلی است. پس توانستیم الگوریتمی
$2-approx$
برای مسئله پوشش راسی ارائه دهیم.
</ul>
</div> 

<div id="section2.3" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <div style="font-size: 17px; color: #ff6600; font-weight: bold">
       راه‌حل با بزرگترین تطابق
    </div>
    <ul>
      در این بخش می خواهیم راه حل جالب دیگری برای تقریب زدن مسئله پوشش راسی ارائه دهیم. برای سادگی در این بخش مسئله پوشش راسی بدون وزن را در نظر می گیریم. در واقع 
        $\forall i: w_{i} = 1$. بنابراین مسئله معادل با یافتن کم ترین تعداد راس می شود که یک پوشش راسی تشکیل دهند. 
        <br> 
        تطابق بیشینه 
        $M$
        را به این صورت می سازیم: 
        ابتدا یکی از یال های گراف را در نظر می گیریم و دو سر آن را در مجموعه ی
        $M$
        قرار می دهیم و این دو راس و یال های متصل بهشان از گراف اصلی حذف می کنیم. سپس از گراف باقی مانده یال دیگری در نظر می گیریم و همین فرآیند را برای تمام یال ها تکرار می کنیم تا در نهایت هیچ یالی در گراف باقی نماند. مجموعه ی به دست آمده مجموعه ی مجزا از یال های گراف است که به آن تطابق بیشینه می گویند.
       <br>  
    برای مثال در شکل ابتدای این بخش یک تطابق بیشینه را به دست آورده ایم.
    <img src="img/maximal.png" width = 400 />
        اکنون ثابت می کنیم 
        $M$
        یک پوشش راسی است.
        فرض کنیم یال 
        $(i, j) \in E$
        وجود داشته باشد که هیچ یک از دو سر آن در 
        $M$
        نباشد. در این صورت
        $M \cup e_{ij}$
        یک تطابق است. (چرا؟)
        اما این با فرض این که 
        $M$
        یک تطابق بیشینه بود در تناقض است. بنابراین هر یالی حداقل یکی از دو سرش در 
        $M$
        است. 
        <br>  
         همچنین به ازای هر 
        $(i, j) \in M$
        هر پوشش راسی 
        $C$
        باید حداقل یکی از 
        $i$
        یا 
        $j$
        یا شامل شود. بنابراین اگر 
        $C^{*}$
        پوشش راسی اپتیمال باشد 
        داریم
        $$|M|  \le 2 \times |C^{*}|$$
        که عبارت بالا از این نتیجه می شود که یال های 
        $M$
        مجزا هستند.
        <br> 
        بنابراین با استفاده از تطابق ماکسیمال نیز می توان یک الگوریتم 
        $2-approx$
        برای مسئله پوشش راسی به دست آورد.
    </ul>
</div>

<div id="section3" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 22px; color: #ff6600; font-weight: bold">
فروشنده دوره‌گرد
</div>
    <div id="section3.1">
    <div style="font-size: 17px; color: #ff6600; font-weight: bold">
        معرفی مسئله
</div>
    <ul>
    مسئله فروشنده دوره‌گرد 
    (Travelling Salesman Problem)
    یا به اختصار TSP
    از مسائل معروف NP
    است که بسیار مطالعه شده‌است.
    در این مسئله تعدادی شهر داریم که نمایانگر رئوس و بین هر دو شهر جاده‌ای داریم که نمایانگر یال‌ها می‌باشند طی کردن هر جاده مدت زمانی طول می‌کشد. حال یک فروشنده دوره‌گرد سعی می‌کند از شهری شروع کند، کل شهر‌ها را یکبار ببیند و سر جای اولیه‌اش برگردد. دغدغه او پیدا کردن چنین دوری است به طوریکه مدت زمانی که در جاده‌ها سپری می‌کند کمینه شود.
    <br>
    به بیان ساده‌تر مسئله فروشنده دوره‌گرد پیدا کردن کوتاهترین دور همیلتونی در یک گراف کامل بدون جهت وزن‌دار با وزن‌های مثبت است.
    <br>
    این مسئله یک مسئله NP-Hard است و نسخه تصمیم‌گیری‌اش به این صورت است:
    <br>
    گرافی وزن‌دار و یک عدد $k$ داده‌شده‌است
    . حال می‌خواهیم ببینیم آیا دور همیلتونی‌ای وجود دارد که وزنش کمتر‌ یا مساوی $k$ باشد.
        </ul>
    </div>
    </div>
</div>    

<div id="section3.2" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <div style="font-size: 17px; color: #ff6600; font-weight: bold">
        راه‌حل حریصانه
    </div>
    <ul>
    حال که با مسئله NP-Hard طرفیم سعی می‌کنیم به سراغ راه‌حل‌های تقریبی برای حل این مسئله برویم.
    برای راحتی کار فعلاً فرض کنید گراف ما اقلیدسی است. یعنی هر رأس معادل نقطه‌ای روی صفحه مختصات است و یال بین دو رأس بیانگر فاصله بین آن دو است.
    <br>
    حال به سراغ حل این حالت می‌رویم. فرض کنید از رأس ۱ شروع می‌کنیم و روشی حریصانه انتخاب می‌کنیم و هر سری بین رئوسی که هنوز ندیدیم رأسی را انتخاب می‌کنیم که فاصله‌اش با یکی از رئوس مجموعه رئوس دیده‌شده کمینه باشد. یعنی یک مجموعه 
    $S$
    داریم که در ابتدا در آن ۱ قرار دارد سپس بین رئوس خارج از 
    $S$
    رأسی را انتخاب می‌کنیم که به $S$ از همه رئوس دیگر نزدیکتر است.
    یعنی وزن مینیمم یالش به این مجموعه کمینه‌است. سپس این رأس را به $S$
    اضافه می‌کنیم 
       فرض کنید در تمامی مراحل می‌خواهیم دوری را نگه داریم و بزرگ کنیم. فرض کنید پس از سه مرحله رئوس $a$ و $b$ و $c$
    به مجموعه اضافه شده‌اند. دور $C$ را برابر این سه‌تایی تعریف می‌کنیم.
    <br>
    سپس در هر مرحله که رأسی مثل $j$ قرار است به مجموعه اضافه شود،
    فرض کنید کوتاهترین یالش به $S$ یالش به رأس $i$ باشد.
    اگر $k$ رأس مجاور $i$ در دور $C$ باشد،
    یال $ik$ را حذف کرده و دو یال $ij$ و $jk$ 
    را به 
    $C$ اضافه می‌کنیم.
    <br>
    حال سراع اثبات این الگوریتم می‌رویم.
    <br>
    الگوریتم در انتخاب کردن رئوس بسیار شبیه به الگوریتم Prim در پیدا کردن درخت کمینه فراگیر عمل می‌کند.
    بنابراین مقایسه وزن کوچکترین درخت فراگیر و کوچکترین دور همیلتونی امری طبیعی است:
    <br>
    <br>
    <span style="color: red">
        لم:
        </span>
        وزن دور همیلتونی کمینه بزرگتر مساوی 
    کوچکترین درخت فراگیر است
    <br>
    اثبات چنین چیزی خیلی راحت است چرا که با حذف یک یال از دور همیلتونی یک مسیر داریم که فراگیر است و چون هر مسیر یک درخت است وزن این مسیر قاعدتاً بیشتر مساوی درخت فراگیر کمینه است در نتیجه وزن این دور بیشتر از درخت فراگیر کمینه است.
    <br>
    <br>
    حالا اگر ثابت کنیم الگوریتم داده‌شده وزنش حداکثر ۲ برابر کوچکترین درخت فراگیر است می‌توانیم به سادگی ثابت کنیم الگوریتم 
    $2-approx$
    می‌باشد
    $$ans \le 2 \times MST \le 2 \times OPT \ \to \ ans \le 2 \times OPT$$
    برای اثبات این موضوع از درستی الگوریتم Prim استفاده می‌کنیم.
    می‌دانیم رأس $ij$ رأسی از درخت فراگیر کمینه برای مجموعه $S$ است.
    اگر وزن $C$ را قبل از تغییرات $c$ بنامیم
    اکنون این وزن اینگونه تغییر کرده‌است:
    $$c' = c - w_{ik} + w_{ij} + w_{jk}$$
    اگر وزن درخت پوشای کمینه در هر مرحله $t$ باشد اکنون به این‌صورت تغییر کرده‌است:
    $$t' = t + w_{ij}$$
    نامساوی مثلثی برای یال‌ها برقرار است یعنی داریم:
    $$w_{jk} \le w_{ij} + w_{ik} \to w_{jk} - w_{ik} \le w_{ij}$$
    بنابراین:
    $$c' = c - w_{ik} + w_{jk} + w_{ij} \le c + 2w_{ij} \le 2t + 2w_{ij} \le 2t' \to c' \le 2t'$$
    در نتیجه الگوریتمی چند جمله‌ای داریم که می‌تواند با ضریب تقریب ۲ مسئله TSP را روی گراف اقلیدسی حل کند!
    به طور دقیق‌تر هر گرافی که نامساوی مثلثی به‌ازای یال‌هایش برقرار باشد این الگوریتم برایش جواب‌گو خواهد بود!
    به این گراف‌ها گراف‌های 
    <strong>متریک</strong>
    می‌گویند.
    <br>
    <br>
    <div style="color: red">
    روش راحتتر برای پیاده‌سازی الگوریتم:
        </div>
    فرض کنید به هر روشی یک درخت پوشای کمینه را پیدا‌کردیم. حالا روی این درخت dfs می‌زنیم و رئوس را به ترتیب دیدنشان کنار هم می‌گذاریم.
    می‌توان ثابت کرد اگر این لیست را به ترتیب بپیماییم و در نهایت به رأس شروعمان برگردیم دور همیلتونی حاصل نیز حداکثر ۲ برابر وزن دور همیلتونی کمینه را دارد.
    </ul>
    </div>
    

<div id="section3.3" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <div style="font-size: 17px; color: #ff6600; font-weight: bold">
        راه‌حلی بهتر
    </div>
    <ul>
    در ادامه به بررسی روشی می‌پردازیم که کمی تقریب بهتری روی گراف‌های متریک به ما می‌دهد.
    کلید این راه‌حل این است که سعی کنیم یک تور اویلری روی گراف پیدا‌کنیم که وزنش کم است. به صورت دقیق‌تر حداکثر
    $\frac{3}{2}$
    وزن کوچکترین درخت فراگیر است.
    <br>
    فرض کنید بازهم کوچکترین درخت فراگیر کمینه را پیدا کردیم، حالا می‌خواهیم تعدادی یال به این درخت اضافه کنیم تا درجه همه‌ رئوس زوج شود، در این‌صورت می‌دانیم تور اویلری‌ای وجود دارد که شامل کل رئوس بشود. برای اینکار رئوس درجه فرد را در مجموعه‌ای مثل $O$ در نظر می‌گیریم.
    حال اگر تطابقی از این رئوس پیدا کنیم و آن‌را به مجموعه یال‌ها اضافه کنیم هر رأسی که قبلا درجه‌اش فرد بوده زوج می‌شود. توجه کنید ممکن است یک یال دوبار تکرار شود در ادامه توضیح می‌دهیم مشکلی بوجود نخواهد آمد.
    <br>
    حالا نکته‌ای که وجود دارد این است که این 
    $\frac{|O|}{2}$
    یال اضافه را طوری انتخاب کنیم که مجموع وزن یال‌ها کمینه شود. برای این‌کار می‌توانیم از الگوریتم کوچکترین تطابق وزن‌دار استفاده نماییم.
    <br>
    حال به لم کلیدی زیر توجه کنید:
    <br>
    <br>
    <span style="color: red;">لم:</span>
    اگر کوچکترین دور همیلتونی کمینه در یک گراف متریک 
    $OPT$
    باشد آن وقت کوچکترین تطابق برای مجموعه 
    $O$
    اندازه‌اش حداکثر
    $\frac{OPT}{2}$
    است.
    <br>
     دور همیلتونی کمینه را در نظر بگیرید و رئوس $O$ را روی آن در نظر بگیرید. اگر این رئوس را به ترتیب شماره گذاری کنیم یکی از دو تطابق 
    $$\{o_1, o_2\}, \{o_3, o_4\}, ..., \{o_{n-1}, o_n\}$$
    یا
    $$\{o_2, o_3\}, \{o_4, o_5\}, ..., \{o_n, o_1\}$$
    وزنش حداکثر 
    $\frac{OPT}{2}$
    است (چرا؟)
<br>
    <br>
    بنابراین حالا که لم ثابت شد می‌توانیم تور اویلری‌ای بیابیم که مجموع وزن یال‌هایش حداکثر 
    $\frac{3}{2}OPT$
    است.
    حالا کافیست در تور اویلری حرکت کنیم و اگر بین رأس $u$ و $v$
    تعدادی رأس تکراری مثل $x_1, x_2,..., x_k$
    وجود داشت آن‌ها را نبینیم و به جای آن مستقیم از $u$ به $v$ برویم.
    به‌دلیل متریک بودن گراف داریم:
    $w_{uv} \le w_{u,x_1} + w_{x_1, x_2} + ... + w_{x_k, v}$
    بنابراین دوری که با انجام این‌کار می‌بینیم وزنی کمتر مساوی تور اویلری دارد که خود کمتر مساوی 
    $\frac{3}{2}OPT$
    است.
</div>

<div id="section4" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 22px; color: #ff6600; font-weight: bold">
سختی تقریب
</div>
    برای تحلیل سختی یک الگوریتم تقریبی با ضریب تقریب 
    $\alpha$
    راه‌های مختلفی وجود دارد. یکی از مرسوم ترین راه‌ها کاهش دادن مسئله به مسئله 
    $NP-Complete$
    است.
    اینطور که می‌گوییم اگر فرض کنیم $P \neq NP$
    در این‌صورت الگوریتمی 
    $\alpha-approx$
    برای مسئله X نداریم.
       <br>
    برای مثال بر می‌گردیم به مسئله فروشنده دوره‌گرد! در حالت متریک الگوریتم‌های تقریبی خوبی به‌دست آوردیم. اما حالا می‌خواهیم ببینیم آیا برای حالت کلی می‌توان برای مثال الگوریتم تقریبی 
    $2-approx$
    ارائه کرد؟
    <br>
    <br>
    <span style="color: red">قضیه:</span>
    اگر $P \neq NP$ در این‌صورت الگوریتمی چند‌جمله‌ای 
    $2-approx$ برای مسئله TSP وجود ندارد.
    <br>
    اثبات: فرض می‌کنیم الگوریتم چند
    جمله‌ای و 
    $2-approx$
    برای این‌کار داریم. با استفاده از این الگوریتم راه‌حلی چند جمله‌ای برای مسئله 
    NP-Complete
    وجود داشتن دور همیلتونی ارائه می‌کنیم.
    به این‌صورت که یال‌های داخل گراف را با وزن ۱ وزن‌دار می‌کنیم و یال‌هاییکه ظاهر نشده‌اند با $n+2$ وزن‌دار می‌کنیم.
    اگر الگوریتم تقریبی را اجرا کنیم و عددی 
    کمتر از $2n+1$ 
    حاصل شد یعنی این گراف دور همیلتونی دارد.
    و در غیر اینصورت فاقد دور همیلتونی است. (توجه کنید این وزن‌گذاری متریک نیست!)
    <br>
    <br>
    می‌توانیم از این هم فراتر برویم فرض کنید الگوریتمی 
    $\alpha-approx$
    برای این مسئله داشته‌باشیم. در این‌صورت اگر وزن یال‌های گراف را برابر ۱ و وزن‌های دیگر را برابر
    $(\alpha - 1) \times n + 1$
    بگذاریم می‌توانیم وجود یا عدم وجود دور همیلتونی را تشخیص دهیم بنابراین. به ازای 
    <strong>
    هر عدد ثابت 
    </strong>
    $\alpha \ge 1$
    <strong>
    هیچ الگوریتم تقریبی چند جمله‌ای 
    </strong>
    $\alpha-approx$
    برای مسئله TSP نداریم!
      <br>
    <br>
    <span style="color: red">تمرین:</span>
    مسئله $k-centre$ مسئله‌ای است که در آن
    تعدادی نقطه در صفحه مختصات داریم. می‌خواهیم این نقاط را در $k$ گروه تقسیم کنیم و برای هر گروه یک نقطه تحت عنوان 
    مرکز بکشیم. وزن هر دسته را برابر ماکسیمم فاصله نقاط آن دسته از مرکزش می‌گذاریم. می‌خواهیم تقسیم بندی و مرکز‌گذاری‌ای را بیابیم به‌طوریکه ماکسیمم وزن مینیمم شود.
    <br>
    الف) سعی کنید الگوریتمی 
    چند‌جمله‌ای
    $2-approx$
     برای این‌کار 
    ارائه دهید.
    <br>
    ب) ثابت کنید هیچ الگوریتم چند جمله‌ای 
    $\alpha-approx, \alpha < 2$
    برای این‌کار وجود ندارد مگر این‌که
    $P = NP$
    <br>
    <br>
    <span style="color: blue">راهنمایی:</span>
    برای بخش ب) سعی کنید با مسئله کوچکترین مجموعه غالب تناظر دهید. مجموعه غالب در یک گراف برابر مجموعه رأسی است که هر رأس در گراف یا داخل مجموعه غالب است یا اینکه به یکی از اعضای مجموعه غالب یال دارد.
</div>
