<div style="font-size: 22px; color: #ff6600; font-weight: bold">
مقدمه
</div><br>
در بخش‌های قبلی درس بررسی کردیم که اگر بخواهیم چند ماتریس را در‌هم ضرب کنیم چگونه بهترین توالی برای این ضرب‌ها را بدست بیاوریم تا سریعتر به جواب برسیم و به روش ضرب دو ماتریس کاری نداشتیم.
<br>
در این بخش چگونگی ضرب دو ماتریس را مورد بررسی قرار می‌دهیم و به دنبال راه‌حلی سریع و بهینه برای این کار هستیم.
</div>
<hr>

<div id="section2" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<h1 style="color: #ff6600;">
بخش اول: الگوریتم استراسن برای ضرب دو ماتریس
</h1><br>            
 دو ماتریس 
$A_{n*n}$
و
$B_{n*n}$
داریم می‌خواهیم 
$C_{n*n} = AB$
را حساب کنیم.
<br>

<hr>
</div>    

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<h2 style="color: #ff6600;">
راه حل اول(راه حل بدیهی)
</h2>         
طبق تعریف ضرب ماتریس‌ها عمل کرده و هر سطر از ماتریس 
$A$
را در هر ستون از ماتریس 
$B$
ضرب می‌کنیم.
<img src="matrix_multiplication.png" style="width:60%">
<strong>
    پیچیدگی زمانی:
</strong>
چون برای محاسبه هر درایه از ماتریس
$C$
باید یک ضرب سطر در ستون انجام شود که هر کدام از مرتبه زمانی 
$O(n)$
زمان می‌برد.
و چون ماتریس 
$C$
دارای
$n^2$
درایه است, انجام ضرب با این روش در مرتبه‌ی زمانی 
$O(n^3)$
انجام می‌شود.
<hr>
</div>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<h2 style="color: #ff6600;">
راه حل دوم (الگوریتم استراسن)
</h2>            
ایده اصلی این الگوریتم تقسیم و حل است. دو ماتریس 
$n*n$
$A$
و
$B$
داده‌شده‌است. با فرض بر اینکه عدد 
$n$
به صورت توانی از عدد
$2$
است.
$(n = 2 ^ k and ‌  k \in N )$
هر کدام از دو ماتریس
$A$
و
$B$
را به 4 زیرماتریس تقسیم می‌کنیم که هر کرام 
$\frac{n}{2} * \frac{n}{2} $
هستند.
(هر کدام از 
$A_{ij}$
ها و
$B_{ij}$
ها ماتریس هستند.)
<br/>


\begin{equation}
A=
\begin{pmatrix}
A_{11} & A_{12} \\
A_{21} & A_{22}
\end{pmatrix}
,
B=
\begin{pmatrix}
B_{11} & B_{12} \\
B_{21} & B_{22}
\end{pmatrix}
\end{equation}


<hr>
</div>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 17px; color: #ff6600; font-weight: bold">
</div>            
به راحتی می‌توان اثبات کرد:
\begin{equation}
C=
\begin{pmatrix}
C_{11} & C_{12} \\
C_{21} & C_{22}
\end{pmatrix}
=
\begin{pmatrix}
A_{11}B_{11} + A_{12}B_{21} & A_{11}B_{12} + A_{12}B_{22} \\
A_{21}B_{11} + A_{22}B_{21} & A_{21}B_{12} + A_{22}B_{22}
\end{pmatrix}
\end{equation}
<img src="strassen_new.png" style="width:50%">

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
در اینجا در هر مرحله باید 8 ضرب ماتریس‌های 
$\frac{n}{2} * \frac{n}{2} $
را حساب کنیم, یعنی:
\begin{equation}
T(n) = 8T(\frac{n}{2}) + O(1)
\end{equation}
که با محاسبه با قضیه اصلی بدست می‌آید:
$T(n) \in O(n^3)$
<br/>
می‌بینیم که زمان اجرای ضرب هنوز بهبود پیدا نکرد. در این روش به‌جای محاسبه این 8 عبارت  
7 عبارت جدید را محاسبه می‌کنیم.
که برای محاسبه هر کدام از این 7 عبارت به محاسبه یک ضرب ماتریس 
$\frac{n}{2} * \frac{n}{2} $
نیاز است.
<br/>
این 7 عبارت به شرح زیر است:
\begin{equation}
M_1 = (A_{11} + A_{22})(B_{11} + B_{22})
\end{equation}
\begin{equation}
M_2 = (A_{21} + A_{22})(B_{11})
\end{equation}
\begin{equation}
M_3 = (A_{11})(B_{12} - B_{22})
\end{equation}
\begin{equation}
M_4 = (A_{22})(B_{21} - B_{11})
\end{equation}
\begin{equation}
M_5 = (A_{11} + A_{12})(B_{22})
\end{equation}
\begin{equation}
M_6 = (A_{21} - A_{11})(B_{11} + B_{12})
\end{equation}
\begin{equation}
M_7 = (A_{12} - A_{22})(B_{21} + B_{22})
\end{equation}
و زیر‌ماتریس‌های 
$C$
به صورت زیر محاسبه می‌شوند:
\begin{equation}
C_{11} = M_1 + M_4 + M_7 - M_5 \\
C_{12} = M_3 + M_5 \\
C_{21} = M_2 + M_4 \\ 
C_{22} = M_1 + M_3 + M_6 - M_1
\end{equation}

<strong>
    نکته:
</strong>
در ابتدای این روش فرض کردیم
$n$
به صورت توانی از 2 است. اگر هم 
$n$
توانی از 2 نباشد با افزودن سطرها و ستون‌های صفر به ماتریس‌ها آنها را بصورت توانی از 2 در می‌آوریم.
<hr>
</div>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<h2 style="color: #ff6600;">
تحلیل زمانی این راه حل
</h2>            
 \begin{equation}
T(n) = 7T(\frac{n}{2}) + O(1)
\end{equation}
<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
که طبق قضیه اصلی:
$T(n) \in O(n^{\lg{7}}) \approx O(n^{2.81})$
</br>
<br>
می‌بینیم که زمان بهتر شد.
</br>
<br>
<strong>
    پرسش:
</strong>
آیا اگر بجای تقسیم ماتریس‌ها به 4 زیر‌ماتریس آنها را به 9 زیرماتریس تقسیم کنیم زمان بهبود پیدا می‌کند؟ 16 زیر ماتریس چطور؟ آیا همیشه با افزایش تعداد زیر‌ماتریس‌ها زمان بهبود میابد؟
<hr>
</div>

<div id="section3" style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<h1 style="color: #ff6600;">
بخش دوم: 
    FFT
    (تبدیل سریع فوریه)
</h1>
</div>    

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    به عنوان آخرین مثال از روش تقسیم و حل، می‌خواهیم به سراغ مسئله معروف و پرکاربردی برویم که شاید قبلا اسم آن را شنیده باشید. تبدیل فوریه!<br>
    البته که کل گستره مبحث تبدیل فوریه خارج از حیطه این درس است، اما می‌خواهیم به نمونه خاصی از حالت گسسته این مسئله بپردازیم که بسیار آشناست.
    می‌خواهیم ضرب دو چندجمله‌ای را به صورت بهینه محاسبه کنیم.
<br>
</div>    

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<h2 style="color: #ff6600;">
تعریف مسئله    
</h2>
    فرض کنید 
    $p(x)$
    و 
    $q(x)$
    دو چندجمله‌ای با درجه
    $n$
    باشند. میخواهیم چندجمله‌ای 
    $p(x).q(x)$
    را محاسبه کنیم.
    <br>
    تلاش اول:
    ساده‌ترین شاید همان روشی باشد که ضرب دو چندجمله‌ای تعریف می‌شود. اگر دو چندجمله‌ای را به شکل  
    $$p(x) = p_n x ^ n + p_{n-1} x ^ {n-1} + ... + p_1 x + p_0 $$
    و
    $$q(x) = q_n x ^ n + q_{n-1} x ^ {n-1} + ... + q_1 x + q_0 $$
    در این صورت ضرب این دو به با
    $n^2$
    عمل ضرب و حدودا
    $n^2$
    عمل جمع قابل انجام است.
    <br>
    تمرین:
    تعداد دقیق ضرب و جمع‌های روش اول برای محاسبه ضرب دو چندجمله‌ای را به دست آوردید و نشان دهید که پیچیدگی زمانی این روش از 
    $O(n^2)$
    است.
    <br>
    حال می‌خواهیم ببینیم چطور می‌توانیم بهتر از این عمل کنیم.
    برای اینکار باید اول نمایشمان از چندجمله‌ای را بازنگری کنیم.
    <br>
    تلاش دوم:
    نمایش یک چندجمله‌ای با ضرایبش را می‌شناسیم، طوری که چندجمله‌ای 
    $$p(x) = p_n x ^ n + p_{n-1} x ^ {n-1} + ... + p_1 x + p_0 $$
    را می‌توان با چندتایی مرتب 
    $$
    (p_n, p_{n-1}, ..., p_1, p_0)
    $$
    به صورت یکتا مشخص کرد. اما راه دیگری هم برای نمایش یک چندجمله‌ای وجود دارد. به عنوان مثال، یک خط با هر دو نقطه روی آن به صورت یکتا تعیین می‌شود.
    یک چندجمله‌ای درجه درجه دو با سه نقطه روی آن به صورت یکتا مشخص می‌شود. به طور کلی هم قضیه زیر برقرار است:
    <br>
    قضیه:
    یک چندجمله‌ای درجه
    $n$
    (با ضرایب مختلط) با
    $n+1$
    نقطه روی آن به صورت یکتا مشخص می‌شود.
    <br>
    دقت کنید که منظور از یک نقطه روی یک چندجمله‌ای مانند
    $p(x)$
    یک زوج مرتب مانند 
    $(x_i, p(x_i))$
    است. برقرار بودن قضیه بالا برای اعداد مختلط نه تنها از نظر ریاضی درست است، بلکه در ادامه برای حل مسئله مورد نظرمان لازم هم خواهد بود.  
    <br>
    با در دست داشتن قضیه بالا، می‌توان نمایش دیگری برای یک چند جمله‌ای پیدا کرد به این صورت که به جای آنکه چندجمله‌ای را با ضرایبش مشخص کنیم، با تعدادی نقطه روی آن مشخص می‌کنیم.
    <br>
    انجام ضرب با این نمایش چند جمله‌ای بسیار ساده است، چون که با داشتن 
    $p(x_i)$
    و
    $q(x_i)$
    می‌توان مقدار
    $p(x_i)q(x_i)$
    را که متعلق به حاصلضرب دو چندحمله‌ای هست را محاسبه کرد.
    <br>
    نکته:
    چندجمله‌ای حاصلضرب یک چندجمله‌ای از درجه
    2n 
    است و باید حاصلضرب 
    2n
    نقطه را به دست آورد.
   <br>
    خب طبق آنچه در بالا گفته شد، محاسبه حاصلضرب دو چندجمله‌ای با 
    2n
    ضرب قابل انجام است.
    اما ما معمولا نمایش عادی یک چند جمله‌ای را داریم، پس سوال این است که هزینه تبدیل چندجمله‌ای به این نمایش چقدر است؟
    شکل زیر خلاصه گفته‌ها تا به اینجا را به صورت خلاصه نشان می‌دهد:
    <img src='fastFourier.png'>
    شکل نشان می‌دهد که روش جدیدی برای محاسبه حاصلضرب پیدا کرده‌ایم اما بهبودی حاصل نشده و پیچیدگی ضرب همچنان 
    $O(n^2)$
    است.
    اگر بتوانیم راهی سریع‌تر برای درونیابی و برونیابی چندجمله‌ای پیدا کنیم می‌توانیم ضرب را سریع‌تر از حالت معمول انجام دهیم.
    این تبدیل نمایش ضرایب به نقاط همان 
    <strong> تبدیل فوریه </strong>
    نام دارد.
</div>    

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <h2 style="color: #ff6600;">
تبدیل فوریه    
</h2>
    طبق آنچه در بالا گفتیم، صورت مسئله ساده شده به این ترتیب است:
    با داشتن نمایش یک چندجمله‌ای درجه 
    n
    با ضرایب به شکل
        $$p(x) = p_n x ^ n + p_{n-1} x ^ {n-1} + ... + p_1 x + p_0 $$
    ، مقدار این چندجمله‌ای را در 
    n
    نقطه به دست آوردید.
    <br>
