```python
#include <iostream>
using namespace std;
const int MAX=1000, MAXINT = 1000000000;

void AllPairsShortestPaths(int G[MAX][MAX], int n) {
  int d[MAX][MAX];
  for (int i = 0; i<n; i++)
    for (int j = 0; j < n; j++)
      d[i][j] = G[i][j];

  for (int k = 0; k < n; k++)
    for (int i = 0; i < n; i++)
      for (int j = 0; j < n; j++)
        if (d[i][k] + d[k][j] < d[i][j])
          d[i][j] = d[i][k] + d[k][j];

  for (int i = 0; i < n; i++)
  {
    for (int j = 0; j < n; j++)
      cout << d[i][j] << "\t";
    cout << endl;
  }
}

int main(int argc, char ** argv){
  int n, m;
  cin >> n >> m;
  int G[MAX][MAX];
  for (int i = 0; i < n; i++)
    for (int j = 0; j < n; j++)
      if (i != j)
        G[i][j] = MAXINT;
      else
        G[i][j] = 0;
  for (int i = 0; i < m; i++)
  {
    int u, v, w;
    cin >> u >> v >> w;
    G[u-1][v-1] = w;
  }
  AllPairsShortestPaths(G, n);
  return 0;
}
```
