    <div>
    تعریف:
        Fibonacci Heap
        یا هرم فیبوناچی یک داده ساختار برای پیاده سازی صف اولویت است. پیچیدگی زمانی هرم فیبوناچی از سایر داده ساختارهای صف اولویت، به طور سرشکن کم‌تر است.

    <div style="font-size: 16px; color: #ff6600; font-weight: bold">
        نکته:
    </div>
        <div style="background-color: lightblue;">
مقادیر این جدول برای هرم فیبوناچی به طور سرشکن تحلیل شده‌اند.
        </div>    
    </div>
    <div style="direction:rtl";>
       ساختار درونی هرم فیبوناچی به شکل اجتماع تعدادی درخت است. تمامی این درخت‌ها هرم کمینه هستند؛ یعنی مقدار هر گره، از مقدار فرزندانش کم‌تر است.
        لازم به ذکر است که این درخت‌ها می‌توانند هر شکلی داشته باشند؛ حتی تمامی این درخت‌ها می‌توانند تنها متشکل از یک راس باشند.
    <br>
     ریشه‌های این درخت‌ها، به وسیله‌ی یک لیست پیوندی دوطرفه
        (doubly)
        و دوری
        (circular)
        به هم متّصل شده‌اند.
        (<a href="https://www.geeksforgeeks.org/doubly-circular-linked-list-set-1-introduction-and-insertion/">doubly circular linked list</a>)
            <br>
        هم‌چنین همواره اشاره‌گری به گره‌ی حاوی عنصر کمینه نگه داشته می‌شود. بدیهی است که چنین گره‌ای، ریشه‌ی یکی از درخت‌هاست.
    </div>
    <img src="img/pic3.png"/>
    <div style="direction:rtl">
        استفاده از لیست پیوندی دوطرفه و مدور دو مزیت دارد؛ اوّل آن که اضافه کردن یک عضو در زمان O(1) انجام می‌شود. دوم آن که اگر دو لیست به ما داده شود می‌توان آنها را در زمان O(1) ترکیب کرد.   
    </div>
    <br>
    <div style="direction:rtl; background-color: lightblue;"> 
         ایده اصلی در هرم فیبوناچی این است که تمامی عملیات را به شکل «lazy»
        انجام دهیم. به طور مثال در عملیات ادغام دو هرم فیبوناچی(merge)
        تنها کاری که انجام می‌دهیم این است که لیست پیوندی شامل ریشه‌های درخت‌های هر دو هرم را به هم متّصل کنیم.
        برای اضافه کردن یک راس به هرم فیبوناچی(insert)
        نیز به سادگی یک درخت تک راس را به مجموعه درخت ها اضافه می‌کنیم.
    </div>
    <br>
    <div style="direction:rtl; background-color: lightyellow;">
    یادآوری: الگوریتم دایسترا(Dijkstra)
    <br>
        فرض کنید یک گراف وزن‌دار G
        داریم. وزن یال‌های این گراف مثبت است.
        هدف الگوریتم دایسترا این است که طول کوتاه‌ترین مسیر از راسی مثل s
        به سایر رئوس گراف را به دست آورد.
    <br>
        در هر مرحله از این الگوریتم باید دو کار انجام دهید:
    <br>
    <ol>
        <li>راس v با کمترین فاصله از S را پیدا کنید</li>
        <li> فاصله‌های کاندید شده برای همه راس‌ها متصل به V را(همسایه های آن)
        آپدیت کنید(فاصله فقط در این مرحله آپدیت می‌شود.)</li>
    </ol>
    اولین مرحله مشابه extract-min 
     کردن در یک صف اولویت است اما برای مرحله دوم چه کنیم؟
    اینجا مفهوم decrease-key وسط می‌آید که کمی مفصل‌تر آنرا بررسی خواهیم کرد.
    </div>
    <br>
    <div style="direction:rtl; background-color: lightyellow;" >
    یادآوری: الگوریتم پریم(Prim)
    <br>
    این الگوریتم درخت پوشای کمینه در یک گراف بدون جهت را پیدا می‌کند.
    <br>
    در هر مرحله از این الگوریتم:
    <ol>
        <li>راس v را پیدا کنید به طوری که اولا (تا این مرحله)
        خارج از درخت ما باشد و کمترین هزینه را برای اضافه شدن داشته‌باشد.
        </li>
        <li>فاصله کاندید شده برای V به راس‌های خارج از ست کنونی درخت </li>
        را بروزرسانی کنید(مشابه مرحله دوم دایسترا)
        </ol>
        بار دیگر مشخص است مرحله اول همان extract-min و مرحله دوم decrease-key خواهد بود.

        پیاده‌سازی
        (Implementation)
        :
    </div>
    <div>
برای پیاده‌سازی این داده‌ساختار، از دو کلاس با interfaceهای زیر استفاده خواهیم کرد. پیاده‌سازی متدهای مربوط به هر بخش، در همان بخش آمده است.
    </div>
    <div>
        در ابتدا کلاس ساده‌ی
        Node
        را داریم که متدهای نسبتا ساده‌ای دارد و نیازی به توضیح اضافه ندارد.
    </div>

```c++
#include <stdlib.h>
template <class V> class FibonacciHeap;
```

```c++
template <class V> struct node {
private:
    /* Linked lists are used to store roots of the heaps
       and also to store children of a node
    */
    /* To handle linked list part */
	node<V>* prev;
	node<V>* next;
    
    /* To save the heap, we need a list of childred and
       a pointer to parent
    */
	node<V>* child;
	node<V>* parent;
    
    /* The value stored in each node */
	V value;
    
    /* Degree of the node */
	int degree;
    
    /* Whether the node is marked or not */
	bool marked;
    
public:
	friend class FibonacciHeap<V>;
	node<V>* getPrev() { return prev; }
	node<V>* getNext() { return next; }
	node<V>* getChild() { return child; }
	node<V>* getParent() { return parent; }
	V getValue() { return value; }
	bool isMarked() { return marked; }

    /* This method is equivalent to `return child != NULL` */
	bool hasChildren() { return child; }
    /* This method is equivalent to `return parent != NULL` */
	bool hasParent() { return parent; }
};
```

    <div>
        در ادامه نیز کلاس
        FibonacciHeap
        آمده است. اکثر پیچیدگی‌های الگوریتم در متدهای این کلاس بیان شده‌اند.
    </div>

```c++
template <class V>
class FibonacciHeap {
protected:
    /* Pointer to the root of a heap with minimum value */
    node<V>* heap;
public:

    FibonacciHeap();
    virtual ~FibonacciHeap();
    
    /* Insert a new value to the FH */
    node<V>* insert(V value);
    
    /* Merge another FH into `this` FH */
    void merge(FibonacciHeap& other);
    
    /* Return true if this FH does not have any node */
    bool isEmpty();
    
    /* Return the minimum value in this FH */
    V getMinimum();
    
    /* Remove the node containing minimum value
       and return its value
    */
    V removeMinimum();
    
    /* Decrease n->value to value */
    void decreaseKey(node<V>* n, V value);
    
    /* Find a node with the given value and return a
       pointer to it
    */
    node<V>* find(V value);
    
private:
    /* Simply return NULL */
    node<V>* _empty();
    
    /* Return a FH which contains a node with the given value */
    node<V>* _singleton(V value);
    
    /* Merge linked lists of a and b
       Then return a pointer to the minimum node
    */
    node<V>* _merge(node<V>* a, node<V>* b);
    
    /* Delete(free space) all nodes */
    void _deleteAll(node<V>* n);
    
    /* Add `child` to the children of `parent` node */
    void _addChild(node<V>* parent, node<V>* child);
    
    /* Set marked=false and parent=NULL for every node in the
       linked list containing node `n`
    */
    void _unMarkAndUnParentAll(node<V>* n);
    
    /* Remove node n */
    node<V>* _removeMinimum(node<V>* n);
    
    /* Remove the edge between node `n` and its parent
       Then this node should be added to `heap` linked list
    */
    node<V>* _cut(node<V>* heap, node<V>* n);
    
    /* Decrease key of node `n` in FH `heap` to the `value` */
    node<V>* _decreaseKey(node<V>* heap, node<V>* n, V value);
    
    /* Find a node with the given value in FH heap */
    node<V>* _find(node<V>* heap, V value);
};
```

```c++
template<class V>
FibonacciHeap<V>::FibonacciHeap() {
    //heap = _empty();
    heap = NULL;
}
```

```c++
template<class V>
FibonacciHeap<V>::~FibonacciHeap() {
    if(heap) {
        _deleteAll(heap);
    }
}
```

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <div>
        همان طور که می‌بینید در پیاده‌سازی هر یک از این دو متد، از یک متد
        private
        استفاده شده است. کد این ۲ متد در ادامه آمده‌اند.
    </div>
</div>

```c++
template<class V>
void FibonacciHeap<V>::_deleteAll(node<V>* n) {
    if(n!=NULL) {
        node<V>* c = n;
        do {
            node<V>* d = c;
            c = c->next;
            _deleteAll(d->child);
            delete d;
        } while(c != n);
    }
}
```

```c++
template<class V>
bool FibonacciHeap<V>::isEmpty() {
    return heap == NULL;
}
```

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <div>
        متد
        _singleton
        یک مقدار دریافت می‌کند و یک هرم تک گره‌ای که می‌سازد که تنها گره آن شامل مقدار ورودی این تابع است.
    </div>

```c++
template<class V>
node<V>* FibonacciHeap<V>::_singleton(V value) {
    node<V>* n = new node<V>;
    n->value = value;
    n->prev = n->next = n;
    n->degree = 0;
    n->marked = false;
    n->child = NULL;
    n->parent = NULL;
    return n;
}
```

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <div>
        متد
        _merge
        دو اشاره‌گر به دو گره دریافت می‌کند. با توجّه به این که این دو گره جزئی از یک لیست پیوندی هستند و البته با این فرض که هر کدام از این ۲ گره در بین اعضای لیست پیوندی خودشان، شامل مقدار کمینه هستند، این دو لیست پیوندی را با یکدیگر ادغام می‌کند و در نهایت اشاره‌گری به گره‌ی کمینه‌ی لیست جدید برمی‌گرداند. بدیهی است که این مقدار بازگشت داده شده، یکی از دو ورودی تابع خواهد بود.
    </div>
</div>

```c++
template<class V>
node<V>* FibonacciHeap<V>::_merge(node<V>* a, node<V>* b) {
    if (a == NULL) return b;
    if (b == NULL) return a;
    if (a->value > b->value) {
        node<V>* temp = a;
        a = b;
        b = temp;
    }
    node<V>* an = a->next;
    node<V>* bp = b->prev;
    a->next = b;
    b->prev = a;
    an->prev = bp;
    bp->next = an;
    return a;
}
```

<div  style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <div>
        در نهایت برای اضافه کردن یک مقدار جدید به کل داده ساختار، ابتدا با استفاده از متد
        _singleton
        یک هرم شامل عنصر جدید ایجاد می‌کنیم.
        سپس تنها گره هرم جدید را با استفاده از متد
        _merge
        به لیست ریشه‌های هرم‌های کنونی داده ساختارمان اضافه می‌کنیم.
    </div>
</div>

```c++
template<class V>
node<V>* FibonacciHeap<V>::insert(V value) {
    node<V>* ret = _singleton(value);
    heap = _merge(heap,ret);
    return ret;
}
```

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <div style="font-size: 16px; color: #ff6600; font-weight: bold">
        اجتماع
        (Merge)
    </div>
    <div>
        برای این قسمت نیز به وسیله‌ی متد
        _merge
        که قبلا پیاده‌سازی کردیم، لیست پیوندی ریشه‌ی هرم‌های دو
        FibonacciHeap
        را با هم ادغام می‌کنیم و در نهایت متغیّر
        heap
        را به ازای هرم دوم برابر با
        NULL
        قرار می‌دهیم.
        <br />
        این کار برای این است که ما در این ادغام از روی داده‌های درخت دوم کپی نمی‌گیریم؛ بلکه عینا خود آن‌ها را به هرم فیبوناچی کنونی منتقل می‌کنیم. در این صورت لازم است که دسترسی به این مقادیر، به هرم فیبوناچی کنونی محدود شود و هرم فیبوناچی دوم نتواند تغییری روی این مقادیر اعمال کند.
    </div>
</div>

```c++
template<class V>
void FibonacciHeap<V>::merge(FibonacciHeap& other) {
    heap = _merge(heap,other.heap);
    other.heap = _empty();
}
```

        دو درخت با رنک برابر k را به شیوه زیر ترکیب می‌کنیم:

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <div style="font-size: 16px; color: #ff6600; font-weight: bold">
        حذف عنصر کمینه
    </div>
    <div>
        همان طور که می‌دانید، این قسمت یکی از پیچیده‌ترین قسمت‌های این داده‌ساختار است! به این منظور، تعدادی تابع کمکی تعریف کرده‌ایم. تابع کمکی اوّل
        _unMarkAndUnParentAll
        است که در واقع روی تمام برادران یک گره به نام
        n
        حرکت می‌کند، در وهله‌ی اوّل
        parent
        آن‌ها را حذف می‌کند و در وهله‌ی دوم
        marked
        را به ازای تمام آن‌ها
        false
        می‌کند.
    </div>
</div>

```c++
template<class V>
void FibonacciHeap<V>::_unMarkAndUnParentAll(node<V>* n) {
    if (n == NULL)
        return;
    node<V>* c = n;
    do {
        c->marked = false;
        c->parent = NULL;
        c = c->next;
    } while(c != n);
}
```

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <div>
        تابع کمکی دیگر، تابع
        _addChild
        است که دو اشاره‌گر به دو گره دریافت می‌کند و یکی از گره‌ها را به لیست بچّه‌های یک گره دیگر اضافه می‌کند.
        <br />
        طرز کار این متد به این شکل است که ابتدا
        next
        و
        prev
        و
        parent
        را به ازای گره فرزند مشخّص می‌کند و پس از افزایش درجه‌ی گره پدر، با متد
        _merge
        که قبلا نوشته بودیم، آن گره را به جمع فرزندان گره پدر اضافه می‌کند.
    </div>
</div>

```c++
template<class V>
void FibonacciHeap<V>::_addChild(node<V>* parent, node<V>* child) {
    child->prev = child->next = child;
    child->parent = parent;
    parent->degree ++;
    parent->child = _merge(parent->child, child);
}
```

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <div>
        سپس متد
        _removeMinimum
        را داریم که یک متد
        private
        برای حذف گرهِ داده شده است.
    </div>
</div>

```c++
template<class V>
node<V>* FibonacciHeap<V>::_removeMinimum(node<V>* n) {
    _unMarkAndUnParentAll(n->child);
    if (n->next == n) {
        n = n->child;
    } else {
        n->next->prev = n->prev;
        n->prev->next = n->next;
        n = _merge(n->next, n->child);
    }
    if (n == NULL) return n;
    node<V>* trees[64] = {NULL};
  
    while(true) {
        if (trees[n->degree] != NULL) {
            node<V>* t = trees[n->degree];
            if(t == n)
                break;
            trees[n->degree] = NULL;
            if (n->value < t->value) {
                t->prev->next = t->next;
                t->next->prev = t->prev;
                _addChild(n, t);
            } else {
                t->prev->next = t->next;
                t->next->prev = t->prev;
                if (n->next == n) {
                    t->next = t->prev = t;
                    _addChild(t, n);
                    n = t;
                } else {
                    n->prev->next = t;
                    n->next->prev = t;
                    t->next = n->next;
                    t->prev = n->prev;
                    _addChild(t, n);
                    n = t;
                }
            }
            continue;
        } else {
            trees[n->degree] = n;
        }
        n = n->next;
    }
    node<V>* min = n;
    node<V>* start = n;
    do {
        if (n->value < min->value) min = n;
        n = n->next;
    } while (n != start);
    return min;
}
```

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <div>
        در نهایت با استفاده از این متدهای کمکی، متد نهایی
        removeMinimum
        که یک متد
        public
        است را پیاده‌سازی می‌کنیم.
        <br />
        در این متد ابتدا عنصر کمینه را ذخیره می‌کنیم. سپس متد
        _removeMinimum
        را به ازای آن گره صدا می‌زنیم. در نهایت حافظه‌ی مربوط به آن گره را آزاد می‌کنیم و مقدار آن گره را برمی‌گردانیم.
    </div>
</div>

```c++
template<class V>
V FibonacciHeap<V>::removeMinimum() {
    node<V>* old = heap;
    heap = _removeMinimum(heap);
    V ret = old->value;
    delete old;
    return ret;
}
```

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <div style="font-size: 16px; color: #ff6600; font-weight: bold">
        کاهش کلید
        (DecreaseKey)
    </div>
    <div>
        ابتدا متدی کمکی به نام
        _cut
        تعریف می‌کنیم. وظیفه‌ی این متد این است که یک گره به نام
        n
        دریافت کند، آن گره را از پدرش جدا کند و سپس آن گره را به یک لیست پیوندی(که در ورودی تابع داده شده) اضافه کند.
    </div>
</div>

```c++
template<class V>
node<V>* FibonacciHeap<V>::_cut(node<V>* heap, node<V>* n) {
    if (n->next == n) {
        n->parent->child = NULL;
    } else {
        n->next->prev = n->prev;
        n->prev->next = n->next;
        n->parent->child = n->next;
    }
    n->next = n->prev = n;
    n->marked = false;
    return _merge(heap, n);
}
```

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <div>
        سپس متد داخلی
        _decreaseKey
        را پیاده‌سازی می‌کنیم.
        وظیفه‌ی این متد این است که مقدار یک گره را به مقدار داده شده «کاهش» دهد.
        گره مذکور، مقدار مذکور و یک لیست پیوندی دو طرفه، ورودی‌های این متد هستند.
    </div>
</div>

```c++
template<class V>
node<V>* FibonacciHeap<V>::_decreaseKey(node<V>* heap, node<V>* n, V value) {
    if (n->value < value)
        return heap;
    n->value = value;
    if (n->parent) {
        if (n->value < n->parent->value) {
            heap = _cut(heap,n);
            node<V>* parent = n->parent;
            n->parent = NULL;
            while (parent != NULL && parent -> marked) {
                heap = _cut(heap,parent);
                n = parent;
                parent = n->parent;
                n->parent = NULL;
            }
            if (parent != NULL && parent->parent != NULL)
                parent->marked = true;
        }
    } else {
        if (n->value < heap->value)
            heap = n;
    }
    return heap;
}
```

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <div>
        در نهایت متد
        decreaseKey
        را پیاده‌سازی می‌کنیم. در این متد تنها کافی است که عضو کمینه‌ی لیست پیوندی ریشه‌های هرم‌ها را به همراه ۲ ورودی دیگر خود این متد، به متد قبلی پاس دهیم.
    </div>
</div>

```c++
template<class V>
void FibonacciHeap<V>::decreaseKey(node<V>* n, V value) {
    heap = _decreaseKey(heap, n, value);
}
```

    <div style="font-size: 16px; color: #ff6600; font-weight: bold">
        پیدا کردن
        (Find)
    </div>
    <div>
        در نهایت، به پیاده‌سازی متدی برای پیدا کردن یک مقدار در بین تمام گره‌ها می‌پردازیم. البته می‌دانیم انجام دادن این کار، لزوما روشی بهینه ندارد و پیچیدگی زمانی‌ش مانند سایر عملیات‌هایی که تا کنون انجام دادیم، خوب نیست!
        <br />
        برای انجام این کار، به نوعی یک الگوریتم
        DFS
        روی تمام هرم‌ها اجرا می‌کنیم تا مقدار مذکور را بیابیم.
        <br />
        لذا در ابتدا یک متد به نام
        find
        می‌سازیم که متد درونی
        DFS
        را صدا می‌زند. اسم این متد درونی را
        _find
        می‌گذاریم.
    </div>

```c++
template<class V>
node<V>* FibonacciHeap<V>::find(V value) {
    return _find(heap, value);
}
```

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
    <div>
        متد
        _find
        به عنوان ورودی یک لیست پیوندی می‌گیرد و یک مقدار.
        قاعدتا این لیست پیوندی توسّط یکی از اعضایش مشخّص می‌شود.
        در این متد ابتدا مطمئن می‌شویم که لیست کنونی خالی نباشد.
        سپس روی تمام اعضای این لیست حرکت می‌کنیم. به هر گره که رسیدیم، ابتدا چک می‌کنیم خود این گره، گره مطلوب ماست یا خیر؛ اگر گره مطلوب ما بود، آدرس آن را برمی‌گردانیم. در غیر این صورت همین تابع را روی لیست بچّه‌هایش اجرا می‌کنیم.
        <br />
        بنابراین تمام گره‌های درخت با الگوریتم
        DFS
        جست‌و‌جو می‌شوند تا به گره‌ای با مقدار مطلوب برسیم.
    </div>
</div>

```c++
template<class V>
node<V>* FibonacciHeap<V>::_find(node<V>* heap, V value) {
    node<V>* n = heap;
    if (n == NULL)
        return NULL;
    do {
        if (n->value == value)
            return n;
        node<V>* ret = _find(n->child,value);
        if (ret)
            return ret;
        n = n->next;
    } while (n != heap);
    return NULL;
}
```

    <div style="font-size: 16px; color: #ff6600; font-weight: bold">
        دایسترا
    </div>
    <div>
        در این مثال نیز ابتدا یک کلاس برای ذخیره‌سازی یک گراف جهت‌دار ایجاد می‌کنیم و سپس روی این گراف با استفاده از الگوریتم دایسترا کوتاه‌ترین مسیر از راسی دلخواه به سایر رئوس را به دست می‌آوریم.
    </div>
</div>

```c++
#define INF (1000 * 1000 * 1000)

typedef pair<int, int> pii;

class Graph {
private:
    vector<pii> *adj;
    int n;
    
public:
    Graph(int n) {
        this->n = n;
        adj = new vector<pii>[n];
    }
    
    ~Graph() {
        delete[] adj;
    }
    
    /* Add an edge from u to v with weight w */
    void addEdge(int u, int v, int w) {
        adj[u].push_back(pii(v, w));
    }
    
    void dijkstra(int source) {
        FibonacciHeap<pii> set;
        node<pii> *pos[n];
        int distance[n];
        
        fill(distance, distance + n, INF);
        distance[source] = 0;
        
        for (int u = 0; u < n; u++)
            pos[u] = set.insert(pii(distance[u], u));
        
        while (!set.isEmpty()) {
            int u = set.removeMinimum().second;
            for (auto e : adj[u]) {
                int v = e.first;
                int w = e.second;
                if (distance[u] + w < distance[v]) {
                    distance[v] = distance[u] + w;
                    set.decreaseKey(pos[v], pii(distance[v], v));
                }
            }
        }
        
        for (int i = 0; i < n; i++)
            cout << distance[i] << " \n"[i == n - 1];
    }
};
```

```c++
{
    Graph g(5);
    g.addEdge(0, 1, 5);
    g.addEdge(0, 2, 10);
    g.addEdge(1, 2, 2);
    g.addEdge(2, 1, 5);
    g.addEdge(2, 4, 3);
    g.addEdge(1, 4, 10);
    g.addEdge(4, 3, 5);
    g.dijkstra(0);
}
```

         <br>
         همانطور که تا به‌حال متوجه شده‌اید هرم فیبوناچی از مجموعه درخت‌های کوچکتر به شکل (heap tree)
        ساخته شده است که در مجموعه از
        محدودیت‌ها قرار دارند در اصل این درخت‌ها به گونه‌ای قرار دارند که درخت با اردر n 
         از حداقل F(n+2)
         راس تشکیل شده باشد که این تابع F
         همان تابع فیبوناچی است.
         <br>
         برای اینکه ببینیم چرا این قضیه همواره برقرار است بیایید بار دیگر به مراحل ساخت این درخت‌ها بازگردیم.در ابتدا هرگاه یک راس به هرم فیبوناچی اضافه می‌شود یک درخت جدید به آن اختصاص داده می‌شود که فقط همان راس را دارد و از اردر صفر است.
         هرگاه یک مقدار نیز حذف می‌شود در نتیجه مراحلی که پیشتر کامل توضیح داده شد برخی از این درخت ‌ها با هم ترکیب می‌شوند که در نتیجه تعداد درخت‌ها بی‌رویه زیاد نشود.
         <br>
         وقتی این درخت‌ها با هم ترکیب می‌شوند همانطور که می‌دانید فقط آنهایی که از یک اردر هستند با هم ترکیب می‌شوند.مراحل ترکیب شدن نیز اینگونه بود که ازمیان دو درختی که درحال ترکیب شدن با هم هستند آن درختی که ریشه 
            بیشتری دارد فرزند مستقیم دیگری می‌شود یکی از نتایج این مرحله این است که در نهایت درخت های از اردر n همواره n فرزند خواهند داشت.
         <br>
         یکی از مواردی که توضیح داده شد این  است که در این داده‌ساختار عمل کاهش‌کلید در زمان سرشکن O(1)انجام می‌شود.
         یکی از مشکلات انجام این کار این است که در عملیات کاهش کلید در عمل فرزندان از پدر جدا می‌شدند پس اگر این اتفاق تعداد زیادی رخ دهد ما در نهایت ممکن است یک درختی با اردر بالا اما تعداد راس بسیار اندک داشته باشیم.
         زمان خوب کارکرد این داده‌ساختار دقیقا وقتی گارانتی می‌شود که ما درخت‌های با اردر زیاد و راس‌های زیاد داشته باشیم.
         <br>
         حال چه‌کار کنیم که این مشکل رخ ندهد؟برای این کار یک قانون قرار داده می‌شود که می‌گوید:اگر شما دو فرزند را از یک درخت بریدید باید در عوض کل آن درخت را از پدرش جدا کنید (اینجا همان مفهوم مارک کردن که در پیاه‌سازی استفاده کردیم نمود پیدا می‌کند)
        .
         <br>
         حال به پاسخ سوال اصلی و دلیل آن نزدیکتر می‌شویم تا به‌حال موارد زیر را فهمیده‌ایم:
         <ul style="margin-right: 0;">
             <li>درختی از اردر n دقیقا n فرزند دارد</li>
             <li>درخت اردر n از ترکیب کردن دو درخت از اردر n-1 و گذاشتن یکی فرزند دیگری بدست می‌آیند</li>
             <li>اگر یک درخت دو فرزندش جدا شوند خودش نیز از پدرش جدا می‌شود</li>
         </ul>
         حال بیاید با مثال و استقرا جلو برویم:
         <br>
         کوچکترین درخت از رتبه 0 تنها یک راس دارد.
         <br>
         کوچکترین درخت از رتبه 1 نیز یک درخت با یک ریشه و یک فرزند است.
         <br>
         اما در مورد رتبه 2 چه می‌توان گفت؟با توجه به قوانینی که بالا گفته شد این درخت حتما دوتا فرزند دارد و از ترکیب دو درخت با رتبه 1 تشکیل شده‌است. در نتیجه درخت در ابتدا دو فرزند دارد یکی از رتبه صفر و دیگری از رتبه یک.
         یه یاد داریم که ما می‌توانیم فرزندان را پس از ترکیب کردن درخت ها از آنها جدا کنیم: در این نمونه ما اگر فرزند درخت با رتبه یک را از آن جدا کنیم ما با یک درخت از رتبه یک باقی می‌مانیم که دو فرزند دارد و هر دو از رتبه صفر هستند. پس کوچکترین یک درخت دودویی با سه راس می‌شود.
         <br>
         حال وارد مرحله استقرا می‌شویم، کوچترین درخت از رتبه n+2 به اینگونه خواهد بود:
         <br>
         اول یک درخت نرمال از رتبه n+2 می‌سازیم که فرزندانی از رتبه
         n+1,n,n-1,...,2,1,0
         دارد.پس از آن این درخت‌ها را به اندازه ممکن کوچک می‌کنیم با بریدن زیر درخت‌ها از آنها به‌گونه‌ای که هیچگاه دو فرزند از یک راس جدا نکنیم. اینکار درختی با فرزندان از رتبه:
         n, n-1,...,1,0,0
         تشکیل می‌دهد.
