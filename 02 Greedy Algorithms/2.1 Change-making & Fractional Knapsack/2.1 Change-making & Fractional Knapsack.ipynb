مقدمه
</div><br>
برای تعریف الگوریتم های حریصانه ابتدا با یک مثال شروع میکنیم که در آن حریصانه عمل کردن یا به عبارتی دیگر در هر مرحله بهترین کار را انجام دادن  به ما جواب بهینه را در یک پروسه میدهد . سپس به معرفی اینکه چه دسته سوالاتی با استفاده از این نوع الگوریتم ها قابل حل هستند و چگونه میتوان اثبات کرد حریصانه عمل کردن بهینه است ، میپردازیم . در ادامه نیز 3 دسته از مسائل معروف این بخش از طراحی الگوریتم را بحث میکنیم . 
<br>
مساله ای که با آن بحث را شروع میکنیم ، مساله قرار دادن $n$ فایل صوتی در یک نوار است . در این مساله هر فایل صوتی با شماره $i$ زمان $L[i]$ دارد و تضمین میشود که جمع زمان همه فایل ها برابر ظرفیت نوار است .  اگر $s(i)$ شماره فایلی باشد که در جایگاه $i$ ام است ،  زمان دسترسی به فایلی در جایگاه $i$
ام را این گونه تعریف میکنیم:
$\sum_{k=1}^{i} L[s(k)]$
هدف پیدا کردن جایگشت $s$ است به طوریکه امید ریاضی دسترسی به یک فایل کمینه باشد .همچنین فرض این است دسترسی به هر فایل احتمال $\dfrac{1}{n}$ دارد. 


<hr>

<div style="direction:rtl; font-size: 17px; font-family: 'XB Zar'; line-height: 200%">
<div style="font-size: 17px; color: #ff6600; font-weight: bold">
 راه حل
</div>         
ابتدا  چیزی که سوال میخواهد را تعریف میکنیم یعنی امید ریاضی دسترسی به یک فایل را تعریف میکنیم . 
$$E[cost(i)] = \sum_{i=1}^{n} \dfrac{cost(i)}{n} $$
$$= \sum_{i=1}^{n}\sum_{k=1}^{i}\dfrac{L[s(k)]}{n} $$
<hr>
<br>
بنابراین هدف کمینه کردن مقدار
$\sum_{i=1}^{n}\sum_{k=1}^{i}L[s(k)]$
است . حال در رویکرد حریصانه میگوییم فایل با کمترین هزینه یا زمان را در جایگاه اول ، فایل با کمترین دومین هزینه را در جایگاه دوم و ... میگذاریم . یعنی در هر مرحله که قرار گرفتیم فایل با کمترین حجم را انتخاب میکنیم . اما  اینکه این رویکرد باعث میشود که مقدار اشاره شده کمینه شود ، باید بررسی شود. خوشبختانه در این مساله به لحاظ شهودی واضح است که رویکرد حریصانه به ما جواب بهینه را میدهد اما اثبات آن به این شکل است که فرض میکیم رویکرد حریصانه جواب بهینه را ندهد یعنی الگوریتم بهینه ای وجود داشته باشد که جایگشت را به صورتی انجام میدهد که $i$ ای وجود دارد و برای آن داریم :
$$L[s(i)] > L[s(i+1)]$$
حال باید نشان دهیم چنین الگوریتمی بهینه نیست و با جابجا کردن مکان $s(i)$ و $s(i+1)$ جواب بهتری بدست می آید 
اگر فایل های این دو مکان را با هم عوض کیم اتفاقی که می‌افتد این است که هزینه برای جایگاه $i$ به اندازه $L[s(i)] - L[s(i+1)]$ کاهش میابد و هزینه برای جایگاه $i+1$ تغییری نمی‌کند . نتیجتا با swap کردن این دو فایل به جواب بهتری رسیدیم پس الگوریتمی که ابتدا فرض کردیم بهینه است، بهینه نیست و از این نتیجه میگیریم هیچ الگوریتم بهینه ای وجود ندارد که $i$ ای باشد که برای آن داشته باشیم $L[s(i)] > L[s(i+1)]$    .پس الگوریتم حریصانه ما تنها الگوریتمی است که این ویژگی را ندارد 
و با تغییر دادن جایگشت در آن همیشه جواب بدتر میشود . در نتیجه
حریصانه در این جا رویکردی است که به ما جواب بهینه میدهد.
</div>

مرتبه زمانی این راه حل 
$O(nlogn)$
است زیرا در ابتدا باید فایل ها را بر اساس مدت زمانشان مرتب کنیم که این کار از مرتبه زمانی $O(nlogn)$ است و سپس هر فایل را در جایگاه خود قرار دهیم .

چگونه اثبات کنیم حریصانه بهینه است ؟
</div>
حل مساله به روش حریصانه زمانی کامل میشود که پس از ارائه روش حریصانه اثبات کنیم این روش بهینه است . به صورت کلی برای اثبات بهینه بودن یک روش حریصانه میتوان به این صورت عمل کرد که هر روش یا الگوریتمی را در نظر گرفت مثلا در مساله ای که در بالا مطرح شده هر جایگشتی را برای فایل ها در نظر گرفت . سپس این جایگشت را به نوعی عوض کرد تا هزینه کاهش یابد . این تغییرات را تا جایی اعمال کرد که دیگر تغییری وجود نداشته باشد که هزینه را کاهش بدهد و نشان داد که جایگشت نهایی همان جایگشت حریصانه ماست . 

<br>
شکل دیگری نیز میتوان به مساله اثبات بهینه بودن روش حریصانه نگاه کرد . به این صورت که در همان مساله بالا هر جایگشت دیگری به جز جایگشت حریصانه را در نظر گرفت و نشان داد هیچ کدام از آنها بهینه نیست نتجتا بهینه ای وجود دارد که روش حریصانه است زیرا هر کدام از جایگشت های دیگر میتوانند بهبود پیدا کنند و یک جایگشت باید باشد که بهبود پیدا نکند .
<br>
این در حالی است که نشان دادن اینکه روش حریصانه بهینه نیست کاری اسان تر بوده و فقط به مثال نقض نیاز دارد . در ادامه با مسائلی که حریصانه راه بهینه یا نابهینه برای آنها ارائه میدهد بیشتر آشنا میشویم

مسئله‌ی اول: زمان بندی
</div><br>
فرض کنید $n$ تسک داریم که هرکدام یک زمان شروع و یک زمان پایان دارند. میخواهیم از این $n$ تسک بیشترین تعداد را انتخاب کنیم به طوریکه هیچ دو کاری همپوشانی نداشته باشند 
<br>        
در روش برنامه نویسی پویا مشاهده خواهید کرد که مساله با مرتبه زمانی $O(n^2)$ قابل حل است و نوع گاه هم به صورت بازگشتی خواهد بود یعنی ابتدا فرض میگوییم اگر تسک اول در زیرمجموعه بهینه وجود دارد  هر تسکی که با تسک اول تلاقی دارد وجود دارد و به 2 مجموعه میرسیم . مجموعه اول شامل تسک هایی که زمان پایانشان از زمان شروع تسک اول زود تر بوده و مجموعه دوم ، تسک هایی که زمان شروعشان از زمان پایان تسک اول دیرتر بوده و مساله تبدیل میشود به این 2 زیر مساله و اگر هم تسک اول در زیرمجوعه بهینه نیست این روند را برای تسک های 2 تا $n$ تکرار میکنیم .
</div>

 رویکرد حریصانه اول
</div> 
با توجه به اینکه صحبت در ارتباط با جزئیات راه حل بالا در این بخش نمی‌گنجد به سراغ حل مساله به روش حریصانه می‌رویم . پس باید معیاری مشخص کنیم که براساس آن تسک ها را مرتب کرده و انتخاب کنیم . 
فرض کنید معیار را زمان شروع تسک ها قرار دهیم  . یعنی تسک‌ها را براساس زمان شروعشان مرتب کرده و به این صورت عمل کنیم که کاری که زودتر از همه آغاز شده را ابتدا انتخاب کنیم و سپس همه کارهایی که با این کار تداخل دارند را حذف کرده و از بین کارهای باقی‌مانده همین پروسه را انجام می‌دهیم تا هیچ کاری باقی نماند. 
<br>
آیا با این رویکرد بیشترین تعداد تسک را انتخاب خواهیم کرد ؟ خیر . فرض کنید کاری که زودتر از همه شروع شده دیرتر از همه تمام شود بنابراین تنها یک کار انتخاب خواهد شد در حالیکه ممکن است بتوانیم چند کار با مدت زمان کمتر را همزمان انتخاب کنیم .
بنابراین روش حریصانه بر اساس زمان شروع بهینه نیست و این مساله را با مثال نقض نشان دادیم

این روش هم روش بهینه نیست زیرا ممکن است کار با کمترین طول با همه کار های دیگر تلاقی داشته باشد در حالی که بتوان چند کار با طول بیشتر را همزمان انتخاب کرد .

 رویکرد حریصانه سوم
</div> 
براساس زمان پایان تسک ها را مرتب کرده و ابتدا تسکی را انتخاب میکنیم که زود تر از همه پایان یابد . سپس تسک هایی که با این تسک تداخل دارن را حذف کرده و این کار را برای تسک های باقی مانده انجام میدهیم .
این روش بر خلاف 2 روش پیشین بهینه بوده و باید آن را اثبات کنیم .
<br>
فرض می‌کنیم روش بهینه دیگری برای انتخاب تسک ها وجود داشته باشد که در آن روش تسک‌های $h_{1}, h_{2}, h_{3}, ..., h_{m}$ به ترتیب به عنوان زیرمجموعه بهینه انتخاب شده است 
همچنین زیرمجموعه ای که توسط الگوریتم حریصانه ما انتخاب شده است به ترتیب  $g_{1}, g_{2}, g_{3}, ..., g_{k}$ است 
حال میخواهیم زیرمجموعه بهینه را به زیرمجموعه حریصانه خودمان تبدیل کنیم .
عضو اول هر دو مجموعه را در نظر میگیریم یعنی $h_1$ و $g_1$ . میگوییم با توجه به اینکه $g_1$ توسط الگوریتم حریصانه انتخاب شده است زمان پایانش کوچک تر مساوی زمان پایان  $h_1$ است و چون $h_1$ عضو زیرمجموعه بهینه است پس با سایر اعضای زیر مجموعه هم پوشانی ندارد . از این 2 گزاره نتیجه میشود که میتوان $h_1$  را حذف و به جای آن $g_1$ گذاشت . زیرمجموعه ساخته شده همچنان بهینه است زیرا هم پوشانی نداشته و سایزش هم عوض نشده است .
بنابراین نشان دادیم هر زیرمجموعه بهینه ای قابل تبدیل به زیرمجموعه ای است که عضو اولش همان عضوی باشد که اولین انتخاب الگوریتم حریصانه است . حال یک حالت کلی را برای این دو زیرمجموعه بررسی میکنیم که $i$ تا عضو اولشان دقیقا یکی است به این صورت :
$$g_{1}, g_{2}, ..., g_{i}, g_{i+1}, ... g_{k}$$
$$g_{1}, g_{2}, ..., g_{i}, h_{i+1}, ...,h_{k}, ..., h_{m}$$
چیزی که تا به اینجا اثبات کردیم برای i=1 
است
میخواهیم حال برای i=2 تا i=k اثبات کنیم . با توجه به اینکه $g_{i+1}$ زودتر از $h_{i+1}$ تمام میشود پس با همه اعضای بعد از $h_{i+1}$ تداخلی ندارد و چون توسط الگوریتم حریصانه ما انتخاب شده با همه اعضای قبل از $h_{i+1}$ نیز تداخل ندارد پس میتوان به جای $h_{i+1}$، $g_{i+1}$ را در زیرمجموعه بهینه قرار داد تا زمانی که i برابر با  k میشود .
حال ادعا میکنیم m  باید حتما برابر با k باشد زیرا اگر هنوز مجموعه ای باشد که دیرتر از $g_{k}$ تمام بشود و با آن تداخل نداشته باشد الگوریتم حریصانه آن را انتخاب میکند و چون حریصانه دیگر عضوی انتخاب نکرده است پس چنین عضوی وجود ندارد 
بنابراین انتخاب بهینه به انتخاب ما توسط الگوریتم حریصانه تبدیل شد.

تحلیل زمانی راه حل سوم
</div>
مرتبه زمانی این راه حل $O(nlogn)$ است ابتدا هر دو آرایه را  مرتب میکنیم .
در هر مرحله عضوی که کمترین زمان پایان دارد را انتخاب میکنیم و هر عضوی که زمان شروع اش کمتر از زمان پایان عضو انتخاب شده است را خط میزنیم و سپس همین کار را تکرار میکنیم . چون هر عضو دقیقا یکبار یا خط خورده است یا انتخاب شده است روند انتخاب اعضا از $O(n)$ است . 
پس در نهایت مرتبه زمانی کل برابر با همان $O(nlogn)$ میشود. در زیر کد این الگوریتم را مشاهده می‌کنید:
</div>

```c++11
#include <bits/stdc++.h> 
using namespace std;
```

```c++11
struct Task 
{ 
    int start, finish; 
};
```

```c++11
bool taskCompare(Task s1, Task s2) 
{ 
    return (s1.finish < s2.finish); 
} 
```

```c++11
void printMaxTasks(Task arr[], int n) 
{ 
    sort(arr, arr+n, taskCompare); 
  
    cout << "Following activities are selected:" << endl; 
  
    int i = 0; 
    cout << "(" << arr[i].start << ", " << arr[i].finish << "), "; 
  
    for (int j = 1; j < n; j++) 
    { 
      if (arr[j].start >= arr[i].finish) 
      { 
          cout << "(" << arr[j].start << ", "
              << arr[j].finish << "), "; 
          i = j; 
      } 
    } 
} 
```

```c++11
Task tasks[] = {{5, 9}, {1, 2}, {3, 4}, {0, 6}, 
                                       {5, 7}, {8, 9}}; 
int numberOfTasks = sizeof(tasks)/sizeof(tasks[0]); 
printMaxTasks(tasks, numberOfTasks); 
return 0;
```

```c++11
#include <bits/stdc++.h> 
using namespace std; 

int deno[] = { 1, 2, 5, 10, 20, 50, 100, 500, 1000 }; 
int numberOfCoins = sizeof(deno) / sizeof(deno[0]); 
```

```c++11
void findMin(int V) 
{ 
    vector<int> ans; 
  
    for (int i = numberOfCoins - 1; i >= 0; i--) { 
        while (V >= deno[i]) { 
            V -= deno[i]; 
            ans.push_back(deno[i]); 
        } 
    } 
  
    for (int i = 0; i < ans.size(); i++) 
        cout << ans[i] << "  "; 
}
```

مسئله‌ی سوم: کوله‌پشتی کسری
</div>
    فرض کنید $n$ شی با ارزش‌های $v_1, v_2, ..., v_n$ و وزن‌های $w_1, w2, ..., w3$ داریم و یک کوله‌پشتی داریم که نهایتا می‌تواند وزن $W$ را تحمل کند. حال می‌خواهیم کوله‌پشتی را با اشیای موجود به نحوی پر کنیم که اگر که در آن $x_i$ ضریبی است که از شی $i$ ام برداشته‌ایم، 
    $$\forall i: 0 \leqslant x_i \leqslant 1$$ و $$\sum_{i=1}^{n}{x_i}{w_i} \leqslant W$$ و مقدار $ \sum_{i=1}^{n}{x_i}{ر_i} \leqslant W$ بیشینه شود. این مسئله به مسئله‌ی کوله‌پشتی کسری معروف است. همین مسئله را اگر تغییر دهیم به نحوی که نتوان کسری از هر شی را برداشت و مجبور باشیم هر شی را برداریم یا نه، به مسئله بسیار سختی تبدیل می‌شود. در قسمت‌های بعدی درس خواهید دید این مسئله که به کوله پشتی صفر و یک معروف است، مسئله‌ای NP-hard است.

 رویکرد حریصانه اول

       این روش به وضوح نابهینه است. فرض کنید چهار شی با وزن‌های ۶۰، ۵۰، ۱۰ و ۴۰ و ارزش‌های ۳۰، ۸۰، ۲۰ و ۱۰ داریم. گنجایش کوله‌پشتیمان هم ۱۰۰ واحد است. با این روش، شی دوم و $\frac{۵}{۶}$ از شی اول را برمی‌داریم و کوله پشتیمان پر می‌شود. با این روش ۱۰۵ واحد ارزش در کوله‌پشتیمان گذاشته‌ایم. در حالی که با یک بررسی ساده و همچنین در رویکرد سوم خواهید دید که بیشتر از این مقدار نیز قابل برداشتن است.

    در این روش چگالی یک شی را $\frac{value}{weight}$ آن شی تعریف می‌کنیم. حال در هر مرحله بین اشیا باقی‌مانده شی با بیشترین چگالی را برمی‌داریم. این عمل را تا جایی ادامه می‌دهیم که کوله پشتی‌مان پر شود. واضح است که در این روش هم ممکن است شی آخر کامل جا نشده و کسری از آن را برداریم. این روش جواب بهینه را به ما می‌دهد. در ادامه به اثبات بهینه بودن آن میپردازیم و درنهایت آن را از نظر مدت زمان اجرا تحلیل می‌کنیم.
    <br>
    ***************************************** proof 1 **********
    <br>
    برای اثبات این رویکرد، فرض می‌کنیم اگر اشیا برحسب چگالی مرتب شده باشند، رویکرد ما $x_i$ و رویکرد دیگری که بهینه است، $x_i^*$ از شی $i$ام برمی‌دارد. شی با بیشترین چگالی را در نظر می‌گیریم و آن را k مینامیم. بدون این شی، کوله‌پشتی در هر دو حالت مقدار یکسانی گنجایش دارد. دو حالت داریم:
    <br>
    1.
    الگوریتم بهینه به همان اندازه ما از آن شی برداشته است. اگر شی دیگری باقی‌نمانده بود، نتیجه می‌شود محتوای دو کوله پشتی با هم برابر است و در نتیجه روش ما بهینه است. در صورتی که شی دیگری موجود بود، همین مراحل را برای شی بعدی با بیشترین چگالی تکرار می‌کنیم.
    <br>
    2.
    الگوریتم بهینه مقدار کمتری از آن شی برداشته است. (واضح است که امکان ندارد مقدار بیشتری برداشته باشد چرا که در الگوریتم ما تا جایی که جا داشتیم شی با بیشترین چگالی را برمی‌داشتیم. اگر شی کامل برداشته‌نشده باشد، کوله پشتی گنجایش نداشته که با توجه به یکسان بودن گنجایش خالی در دو حالت، در رویکرد بهینه هم جا نخواهد داشت.) حال با توجه به آنکه بدون این شی کوله‌پشتی‌ها جای خالی یکسانی دارند، میتوان مقدار $x_k - x_k^*$ از این شی به کوله پشتی بهینه اضافه کرد و از بقیه موارد کوله پشتی بهینه (که مقدار چگالی کمتری دارند) برداشت. واضح است که این کار مقدار ارزش کوله را کمتر نمی‌کند. (در صورتی که چگالی جسمی که برداشته شده با گذاشته شده برابر باشد، ارزش کوله تغییری نمی‌کند و در غیر این صورت کمتر می‌شود.) در صورتی که ارزش کوله کمتر شود با بهینه بودن کوله‌پشتی بهینه به تناقض می‌خوریم و درصورتی که ارزش آن تغییری نکند، یک مرحله به کوله پشتی ما شبیه‌تر شده است و همین مراحل را برای شی بعدی با بیشترین چگالی تکرار می‌کنیم.
    <br>
    بدین ترتیب کوله پشتی بهینه بدون تغییر ارزش به کوله پشتی ما تبدیل شد. بنابراین کوله پشتی ما نیز بهینه بوده‌است و این رویکرد، کوله پشتی بهینه را به ما می‌دهد.

 تحلیل زمانی رویکرد سوم

مرتبه زمانی این راه حل $O(nlogn)$ است. ابتدا اشیا را برحسب چگالی‌شان را  مرتب میکنیم.
در هر مرحله عضوی که کمترین چگالی را دارد انتخاب می‌کنیم و سپس سراغ شی بعد می‌رویم تا جایی که کوله‌پشتی‌مان پر شود. روند انتخاب اشیا از $O(n)$ است . 
پس در نهایت مرتبه زمانی کل برابر با همان $O(nlogn)$ که زمان مرتب کردن اشیا بود می‌شود. در زیر کد این الگوریتم را مشاهده می‌کنید:

```c++11
#include <bits/stdc++.h> 
using namespace std; 
  
struct Item 
{ 
    int value, weight; 
  
    Item(int value, int weight) : value(value), weight(weight) 
    {} 
};
```

```c++11
bool cmp(struct Item a, struct Item b) 
{ 
    double r1 = (double)a.value / a.weight; 
    double r2 = (double)b.value / b.weight; 
    return r1 > r2; 
}
```

```c++11
double fractionalKnapsack(int W, struct Item arr[], int n) 
{ 
    sort(arr, arr + n, cmp); 
    int curWeight = 0; 
    double finalValue = 0.0; 
  
    for (int i = 0; i < n; i++) 
    { 
        if (curWeight + arr[i].weight <= W) 
        { 
            curWeight += arr[i].weight; 
            finalValue += arr[i].value; 
//             cout << "pick all of {" // if you want to see result items
//                 << arr[i].value << ", " 
//                 << arr[i].weight << "}" << endl;
        } 
        else
        { 
            int remain = W - curWeight; 
            finalValue += arr[i].value * ((double) remain / arr[i].weight); 
            if (remain != 0){
//                 cout << "pick " << ((double) remain / arr[i].weight) // if you want to see result items
//                     << " of {" << arr[i].value 
//                     << ", " << arr[i].weight << "}" << endl;
            }
            break; 
        } 
    } 
  
    return finalValue; 
} 
```

```c++11
int W = 50; 
Item items[] = {{60, 10}, {100, 20}, {120, 30}}; 

int numberOfItems = sizeof(items) / sizeof(items[0]); 

int maxvalue = fractionalKnapsack(W, items, numberOfItems);
  
cout << "Maximum value we can obtain = "
     << maxvalue; 
```
